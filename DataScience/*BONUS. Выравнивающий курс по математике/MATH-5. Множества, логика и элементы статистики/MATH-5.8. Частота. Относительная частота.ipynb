# <center> MATH-5.8. Частота. Относительная частота

***Частотная таблица*** — это простой способ представления и обобщения большого количества чисел.

Чтобы построить частотную таблицу, необходимо выстроить от меньшего к большему все возможные значения, которые встречаются в данных, а затем для каждого их них высчитать частоту — то, сколько раз встречается это значение.

Например, если мы хотим построить распределение оценок (по пятибалльной шкале), которые выставили нашему товару покупатели на маркетплейсе, то у нас будет пять вариантов значений: 1, 2, 3, 4, 5. Для каждого значения мы можем посчитать, сколько раз такая оценка была выставлена. По полученному распределению можно сразу оценить основную тенденцию: каких оценок большего всего, часто ли клиенты недовольны нашей продукцией, каково соотношение довольных и недовольных покупателей и т. д.

Давайте рассмотрим наш пример с оценками не в теории, а на практике. Пусть у нас есть 25 клиентов, которые купили товары и поставили им оценки от 2 до 5:

3, 2, 3, 3, 5, 4, 3, 2, 4, 2, 3, 4, 5, 2, 4, 3, 4, 5, 5, 4, 3, 3, 2, 3, 2

Даже такой небольшой набор данных довольно сложно воспринимать без группировки. Если посчитать, сколько раз каждая оценка встречается в этом наборе чисел, получим, что двойка встречается пять раз, тройка — десять раз, четвёрка — шесть раз и пятёрка — четыре раза.

Оформим это в виде таблицы и получим частотное распределение:

- Значение - Частота
- 2 - 5
- 3 - 10
- 4 - 6
- 5 - 4

Глядя на такую таблицу, мы сразу можем сделать ряд выводов: например, о том, что отличных оценок очень мало, а большинство клиентов скорее недовольны нашими товарами.

С помощью такого распределения можно так же быстро сделать дополнительные вычисления: для начала мы можем посчитать сумму всех оценок, но не складывая все оценки по очереди, а складывая произведения значения и его частоты.

$2*5+3*10+4*6+5*4=84$

Так же просто можно посчитать среднее значение, если поделить полученную сумму на количество значений:

$\frac{84}{25}=3.36$

Иногда в распределениях считают не абсолютные частоты, а пропорции или проценты. Так проще показать, какую долю (или же относительную частоту) занимают те или иные значения. В нашем примере мы видим, что двойку поставили 5 человек из 25. Значит, мы можем рассчитать, что это $\frac{5}{25}=0.2$ от общего количества (если считать долю) или $0.2*100=20$ (если считать в процентах).

Если значений не очень много, такое представление довольно удобно. Однако при большом количестве чисел таблица получится очень объёмной и делать по ней выводы будет сложно. Для таких случаев можно строить ***распределение частот с группировкой***: считать частоты не для конкретного значения, а для их группы. Это особенно актуально для количественных переменных.

Например, пусть у нас есть данные, которые содержат баллы за ЕГЭ по информатике для группы школьников:

45, 55, 20, 63, 65, 34, 55, 88, 73, 78, 75, 20, 98, 63, 93, 60, 14, 88, 90, 75, 95, 73, 34, 27, 100

Если мы будем выписывать частоту для каждого значения, таблица получится громоздкой, поэтому сгруппируем значения в пять групп одинакового размера: 1–20 баллов, 21–40 баллов, 41–60 баллов, 61–80 баллов, 81–100 баллов. Получаем следующее распределение:

- Значение - Частота
- 0-20 - 3
- 21-40 - 3
- 41-60 - 4
- 61-80 - 8
- 81-100 - 7

Разумеется, так анализировать распределение стало удобнее. Однако стоит отметить, что в таком виде оно потеряло часть информативности. Например, мы знаем, что восемь человек получили баллы в интервале от 61 до 80, но совершенно не ясно, какие именно это баллы — выше 70 или ниже? Может быть, они все близки к нижней границе, а может быть — к верхней. Такие уточнения могут быть полезны для формулирования выводов, однако после группировки мы теряем эту информацию. Поэтому всегда стоит анализировать, подходит ли в конкретном случае частотное распределение с группировкой.

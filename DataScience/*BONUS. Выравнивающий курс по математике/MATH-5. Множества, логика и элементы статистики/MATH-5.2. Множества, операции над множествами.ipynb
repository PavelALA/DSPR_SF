# <center> MATH-5.2. Множества, операции над множествами


Обычно под ***множеством*** понимают объединение (семейство, класс) объектов, обладающих каким-то общим признаком или совокупностью признаков.

Множество состоит из ***элементов***. Их может быть сколько угодно, вплоть до бесконечности. Важно, что все элементы множества уникальные.

Множество, не имеющее элементов, называется ***пустым*** и обозначается как Ø или {}.

Самый простой способ задать множество — просто перечислить его объекты. Например, мы можем создать множество с навыками, полезными для аналитика:

W={'математика','программирование','базы даных'}

Мы можем обозначить, что элемент входит во множество:

$'математика'\in W$

('математика' входит в множество W)

Если элемент не входит во множество, используется тот же символ в перечёркнутом виде:

$'история средних веков'\notin W$

('история средних веков' не входит в множество W)

## <center> ОПЕРАЦИИ НАД МНОЖЕСТВАМИ И СВЯЗАННЫЕ С НИМИ ОБОЗНАЧЕНИЯ

### ОБЪЕДИНЕНИЕ

***Объединением*** множеств A и B называют множество C, состоящее из элементов, принадлежащих множеству A или множеству B (или обоим этим множествам сразу):

$C=A\cup B$

Таким образом, во множество C помещаются абсолютно все элементы из двух множеств (но ***без повторений***).

Схема ниже наглядно демонстрирует объединение множеств A и B. На этой и подобных схемах ниже нас будет интересовать только закрашенная область.

Представим, что A — это множество людей, которые знают французский язык, а B — множество людей, которые знают немецкий. Тогда объединение A и B — это все люди, которые знают хотя бы один из этих языков (или оба сразу).

### ПЕРЕСЕЧЕНИЕ

Пересечением множеств A и B называют множество C, состоящее из элементов, которые являются общими для множеств A и B:

$C=A\cap B$

### РАЗНОСТЬ

***Разностью*** множеств A и B называют множество C, состоящее из таких элементов множества A, которые не являются элементами множества B:

$C=A\setminus B$

### СИММЕТРИЧЕСКАЯ РАЗНОСТЬ

Симметрическая разность двух множеств — это группа всех тех объектов, которые принадлежат только одному из множеств:

$C=A\bigtriangleup B$

### ДОПОЛНЕНИЕ

Дополнением множества A называют множество $\overline A$, которое является разностью универсального множества U и множества A:

$\overline A=U \setminus A$

***Универсальное множество*** в математике — это совокупность всех существующих в мире объектов. То есть если мы ищем дополнение ко множеству, то получаем абсолютно все объекты, кроме элементов данного множества.

Зачастую под универсальным множеством понимают не вообще все возможные объекты, а объекты в контексте рассматриваемой задачи. Например, это могут быть все люди. Тогда, если во множестве A находятся люди, знающие немецкий, то в дополнении ко множеству A будут все остальные люди, то есть те, кто не знаком с немецким.

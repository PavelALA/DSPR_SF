# <center> MATH-5.7. Размах. Меры разброса

Давайте рассмотрим на примере, зачем это может быть нужно. Например, мы узнаём, что средняя оценка, которую выставили клиенты нашим товарам на маркетплейсе, равна 5 (по десятибалльной шкале). Это может означать, что все клиенты постоянно ставят 5. Также это может значить, что половина клиентов были крайне недовольны и выставили 0, а половина, наоборот, были очень довольны, и выставили 10. В таком случае среднее тоже будет равно 5. Однако эти ситуации совершенно разные. Таким образом, только по характерному значению нельзя составить объективное представление о распределении значений.

## <center> РАЗМАХ И МЕЖКВАРТИЛЬНЫЙ РАЗМАХ

***Размах*** — это разность между наибольшим и наименьшим значениями результатов наблюдений.

Для того чтобы вычислить размах, необходимо вычесть минимум из максимума.

Например, у нас есть следующий ряд чисел:

2,3,6,1,7

Самое большое значение в этом ряду равно 7, а самое маленькое — 1. Значит, размах равен 7-1=6.

У размаха есть тот же недостаток, что и у среднего арифметического — чувствительность к выбросам. Поэтому мы можем использовать другую мер разброса — ***межквартильный размах***.

Снова рассмотрим ряд 2,3,6,1,7. Для начала расположим числа по возрастанию:

1,2,3,6,7

Далее найдём медиану — она равняется 3. Медиана делит ряд на два ряда: 1,2 и 6,7.

Найдём медиану первого ряда — она будет равна 1.5. Это ***нижний квартиль*** (обычно обозначается как $Q_1$). Далее найдём медиану второго ряда — она равна 6.5. Это ***верхний квартиль*** (обычно обозначается как $Q_3$). Теперь из 6.5 вычитаем 1.5 и получаем 5. Это и есть ***межквартильный размах***.

Всего в данных может быть три квартиля, и они делят все объекты на четыре равные части (поэтому и называются ***квартилями***). Ниже представлен пример того, как квартили разбивают данные на равные наборы:

## <center> ДИСПЕРСИЯ. СТАНДАРТНОЕ ОТКЛОНЕНИЕ

Существует ещё две взаимосвязанных меры разброса — дисперсия и стандартное отклонение. Они понадобятся нам, чтобы оценить, насколько в среднем значения разбросаны относительно среднего арифметического.

Важно отметить, что, поскольку для вычисления этих мер используется среднее, они также будут чувствительны к выбросам.

Пусть у нас есть ряд 1,2,3.

Что получить дисперсию, в первую очередь найдём среднее арифметическое:

$\frac{1+2+3}{3}=2$

Теперь вычтем из каждого значения среднее и возведём полученные значения разницы в квадрат:

- Значение: 1,2,3
- Разница значения и среднего: 1-2=-1, 2-2=0, 3-2=1
- Квадрат разницы: 1, 0, 1

Итак, мы понимаем, насколько сильно каждое из значений отличается от среднего. Однако каждое отклонение в отдельности нам не очень интересно — мы хотели бы найти обобщённую оценку.

Для этого сложим возведённые в квадрат значения разниц и поделим результат на их количество:

$\frac{1+0+1}{3}=\frac{2}{3}$

Примечание. На самом деле в статистике при поиске дисперсии обычно делят не на количество наблюдений, а на ***количество наблюдений - 1***: считается, что если мы рассматриваем выборку, то так значение будет точнее. Однако здесь мы опустим этот момент — в тонкостях вы разберётесь далее в соответствующих модулях курса.

Итак, мы получили дисперсию. Чтобы получить стандартное отклонение, надо извлечь корень из дисперсии

То есть стандартное отклонение равно: $\sqrt{\frac{2}{3}}$

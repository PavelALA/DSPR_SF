# <center> MATH-5.6. Среднее арифметическое, мода и медиана

Мы уже знаем, что цель аналитика или исследователя — на небольшой выборке оценить происходящее в генеральной совокупности и найти решение какой-то задачи, например оценить минимальную зарплату, которую получил выпускник курсов по Data Science. То есть нас интересуют не отдельные результаты для каждого объекта, а некая общая оценка, которая позволит дать обобщённую интерпретацию. Чтобы получить такую оценку, в первую очередь необходимо узнать, какие значения являются наиболее типичными для этого признака, что для него наиболее характерно. Для этого мы будем вычислять ***меры центральной тенденции.***

## <center> СРЕДНЕЕ АРИФМЕТИЧЕСКОЕ

Мера центральной тенденции, которая сразу приходит на ум, — ***среднее арифметическое***.

Вычислять среднее арифметическое довольно просто: надо просто сложить все показатели и разделить получившуюся сумму на их количество. Например, мы хотим посчитать среднюю зарплату для трёх человек: зарплата первого — 45 000, зарплата второго — 70 000, зарплата третьего — 65 000. Тогда средняя зарплата будет равняться:

$\frac{45000+70000+65000}{3}= 60000$

Однако у среднего арифметического есть существенный минус — ***эта мера чувствительна к выбросам***.

Есть несколько способов справиться с этой проблемой:

1. Вычислять ***усечённое среднее***.

Чтобы его найти, необходимо отбросить самые маленькие и самые большие значения (например, по 5 % с каждой стороны) и вычислять среднее уже по такой урезанной выборке. Отсекая самые маленькие и самые большие значения, мы избавляемся от выбросов, и результат получается более релевантным.

2. Вычислять ***показатели***, ***не чувствительные к выбросам.***

Аналог среднего арифметического, который не чувствителен к выбросам, — это медиана.

## <center> МЕДИАНА

***Медиана*** — это число, которое находится в середине набора чисел, если его упорядочить по возрастанию.

Например, у нас есть ряд чисел:

1,6,2,7,3

Выстроим их по возрастанию

1,2,3,6,7

Найдём центральное значение — в этом ряду это 3. Это и будет медианой.

Если в ряду чётное количество объектов, одного центрального показателя не будет. В таком случае нужно взять два центральных показателя и найти их среднее арифметическое.

## <center> МОДА

Чтобы вычислить ***моду***, необходимо просто найти значение, которое встречается чаще всего.

К примеру, в ряду 1,3,5,3,6,2,2,3 модой будет 3, так как число 3 встречается наибольшее количество раз.

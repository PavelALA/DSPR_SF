# <center> MATH-5.3. Мощность множества. Булеан. Счётные и несчётные множества

## <center> МОЩНОСТЬ МНОЖЕСТВА

***Мощностью*** множества называется количество ***уникальных*** элементов, которые содержатся в определённом множестве. Её обозначают вертикальными линиями, как и модуль.

Например, мощность множества A={1,5,7,8} равна $|A|=4$. Для множества B={4,7,7,9,5} мощность также будет равна $|B|=4$, так как мощность множества составляют только уникальные элементы.

Может показаться, что мощность — довольно простая характеристика, которая не даёт никакой особо важной информации. Однако это не так. Чтобы в полной мере оценить полезность мощности множества, необходимо рассмотреть ещё одно понятие — подмножество.

Например, пусть у нас есть множество W, состоящее из чисел {1,2,3,4}, и множество B, состоящее из чисел {2,3}. Абсолютно все элементы, которые находятся во множестве B, есть и во множестве W. Поэтому мы можем сказать, что множество 
В является подмножеством множества W.

Это можно обозначить это следующим образом:

$B\subset W$ («множество B является подмножеством множества W»)

Разумеется, может быть и такое, что одно множество не является подмножеством другого. Мы можем придумать ещё одно множество C, не все элементы которого являются элементами W:

C={4,5,6,7}

Такое множество C не является подмножеством W. Это обозначается так:

$C\nsubseteq W$

Важно обозначить, что для любого множества пустое множество и само это множество также являются подмножествами. То есть, например, для множества {1,3,5} все возможные подмножества будут следующими:

{Ø},{1},{3},{5},{1,3},{1,5},{5,3},{1,3,5}

Как понять, сколько всего подмножеств существует у конкретного множества? Тут как раз и помогает мощность: она позволяет оценить количество всех возможных подмножеств, существующих в данном множестве.

Чтобы найти количество всех существующих подмножеств, необходимо просто возвести 2 в степень, равную мощности множества — n:

$N_подмножеств=2^n$

## <center> БУЛЕАН

Обычно булеан для множества A обозначается как S(A).

Например, если у нас есть множество A={1,5,8}, то в его булеане будут содержаться следующие элементы:

- {}
- {1}
- {5}
- {8}
- {1,5}
- {1,8}
- {5,8}
- {1,5,8}

Перечислив все множества, входящие в S(A), мы можем ещё раз убедиться, что их количество равняется $2^n=a^3=8$

## <center> СЧЁТНЫЕ И НЕСЧЁТНЫЕ МНОЖЕСТВА

Бывают конечные и бесконечные множества. Если, к примеру, мы рассматриваем множество натуральных чисел, то в нём бесконечное количество элементов.

Однако бесконечные множества тоже бывают разными: они делятся на счётные и несчётные.

***Счётное бесконечное множество*** — это бесконечное множество, элементы которого возможно пронумеровать натуральными числами.

Например, целые числа {...,-3,-2,-1,0,1,2,3,...} являются счётным множеством.

Ниже можно увидеть сопоставление целых чисел с натуральными:

- $1\rightarrow 0$

- $2\rightarrow 1$

- $3\rightarrow -1$

- $4\rightarrow 2$

- $5\rightarrow -2$

- $6\rightarrow 3$

- $7\rightarrow -3$

- $8\rightarrow 4$

- ...

Список бесконечен и содержит все натуральные числа в левой части и все целые числа — в правой.

Таким образом, множества натуральных и целых чисел имеют одинаковое количество элементов (одинаковую мощность). Это может показаться абсурдным и идти вразрез со здравым смыслом, однако это действительно так с точки зрения математики: целых чисел столько же, сколько натуральных, так как можно задать принцип, по которому мы будем сопоставлять их.

***Примечание***. Соотношение целых и натуральных чисел можно записать как единое правило:

$n\rightarrow 2n$, для n>1

$n\rightarrow -2n+1$, для $n\leq1$

Рациональные числа также являются счётным множеством. Для лучшего понимания можно представить это в виде таблицы:

В ней есть все положительные рациональные числа. Они расположены таким образом, что можно двигаться по стрелкам, начиная с левого верхнего угла, составить все дроби с помощью натуральных чисел и таким образом показать, что множество является счётным. Обратите внимание, что по строкам располагаются значения числителя, а по столбцам — знаменателя.

***Несчётное множество*** — это бесконечное множество, не являющееся счётным.

Вещественные числа являются несчётным множеством, так как нет возможности сделать однозначное соотношение всех вещественных чисел к натуральным.

Также несчётными множествами являются все точки числового отрезка: например, если рассматривать все числа на промежутке от 7 до 9 включительно, невозможно сформулировать правило, по которому их можно будет пересчитать с помощью натуральных чисел.

Кроме того, несчётным является множество всех бинарных (состоящих из 0 и 1) последовательностей любой длины. Как бы мы ни пытались пронумеровать такие последовательности, в них можно до бесконечности увеличивать число элементов и сочинять всё новые последовательности.

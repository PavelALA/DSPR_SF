# <center> MATH-5.4. Высказывания и логические операции. Булевы функции

## <center> ВЫСКАЗЫВАНИЯ И ЛОГИЧЕСКИЕ ОПЕРАЦИИ

***Математическая логика*** — это раздел математики, изучающий способы логических рассуждений, доказуемость логических рассуждений и т. п. К математической логике относится множество разделов, которые могут изучаться и в других разделах математики, например теория алгоритмов и теория вычислимости.

Одним из центральных понятий в математической логике является логическое высказывание.

***Логическое высказывание*** — это повествовательное предложение, о котором можно однозначно сказать, истинно оно или ложно.

## <center> ИНВЕРСИЯ

***Инверсия*** — это логическая операция, которая преобразует исходное высказывание в новое, значение которого противоположно исходному.

То есть если высказывание A истинно, то инверсия для высказывания A ложна, и наоборот.

## <center> КОНЪЮНКЦИЯ

***Конъюнкцией*** в математической логике называется логическое «И». Высказывание «A и B» истинно только в том случае, если оба эти высказывания истинны одновременно.

Если высказывание  A— это утверждение «Мария решила все задачи в модуле по математике», а высказывание B — это утверждение «Пётр решил все задачи в модуле по математике», то конъюнкция $A\land B$ будет истинной только случае, если и Мария, и Пётр решили все задачи в модуле по математике.

## <center> ДИЗЪЮНКЦИЯ

Дизъюнкцией называется логическое «ИЛИ». Высказывание «A или B» истинно только тогда, когда истинно хотя бы одно из них (или оба сразу).

Если вновь обратиться к примеру с Марией и Петром, то дизъюнкция $A\lor B$ будет истинной в том случае, если Мария выполнила все задания по математике или Пётр выполнил все задания по математике или они оба их выполнили.

## <center> БУЛЕВЫ ФУНКЦИИ

***Булева функция*** — это функция, в качестве аргументов которой выступают какие-то логические высказывания, а значением может быть только 0 или 1.

# <center> MATH-5.9. *Бонус: метод математической индукции

***Математическая индукция*** — это особый способ доказательства, состоящий из двух шагов:

- Доказать, что утверждение верно для первого элемента (***база индукции***).
- Доказать, что если утверждение верно для какого-то случайного элемента (***индукционное предположение***), то оно верно для всех элементов (***индукционный переход***).

Принцип доказательства с помощью математической индукции работает по принципу домино: каждая падающая костяшка толкает следующую. То есть из верности утверждения для первого элемента следует верность для второго, из верности для второго следует верность для третьего и так далее.

Приведём пару наглядных примеров:

- По правилам этикета мужчина не может стоять в очереди перед женщиной (он должен пропустить её вперёд). Поэтому если первый человек в очереди — мужчина, то за ним тоже должен стоять мужчина. Получается, что второй человек — мужчина, и за ним тоже должен быть мужчина. Так можно продолжать до бесконечности и доказать, что если первый человек в очереди — мужского пола, то и остальные люди в очереди — мужчины.
- Также с помощью индукции можно доказать, что в любой автобус может поместиться любое число людей. Действительно, один человек точно сможет поместиться в автобус (базис индукции). С другой стороны, как бы сильно ни был наполнен автобус, один человек в него всегда влезет (шаг индукции).

Если говорить о числах, то доказательство строится следующим образом:

1. Доказать, что утверждение верно для наименьшего числа (обычно для 1).
2. Доказать, что если утверждение верно для $n=k$, то оно верно и для $n=k+1$.


***ПРИМЕР 1***

Докажем, что выражение $3^n-1$ всегда кратно 2.

1. Проверим, верно ли это утверждение для $n=1$:

$3^1-1=2$

Разумеется, 2 делится на 2, так что утверждение верно.

2. Предположим, что это утверждение верно для $n=k$. Тогда $3^k-1$ делится на 2.

Теперь докажем, что для $n=k+1$ это тоже будет верно. Для этого вместо k подставим k+1 — получаем:

$3^{k+1}-1$

Воспользуемся свойствами степеней и преобразуем выражение:

$3^{k+1}-1=3*3^k-1$

Теперь надо как-то получить в этом выражении , для которого нам известна делимость. Можно сделать это следующим образом:

$3*3^k-1=3*3^k-1-2+2=3*3^k-3+2=3(3^k-1)+2$

Так как $3^k-1$ делится на 2 (т. е. является чётным), то выражение $3(3^k-1)$ тоже является чётным. Конечно, если прибавить к чётному выражению 2, оно останется чётным. Таким образом мы доказали, что $3(3^k-1)$ всегда кратно 2.



***ПРИМЕР 2***

Попробуем доказать с помощью метода математической индукции следующее тождество:

$1+2+3+...+n=\frac{n(n+1)}{2}$

1. База индукции: проверим равенство тождества для $n=1$:

$\frac{n(n+1)}{2}=\frac{1(1+1)}{2}=1$

Видим, что для $n=1$ тождество выполняется.

2. Теперь сформулируем индукционное предположение. Пусть наше равенство верно для какого-то значения $n=k$:

$1+2+3+...+k=\frac{k(k+1)}{k}$

Теперь попробуем доказать, что в таком случае утверждение верно и для $n=k+1$. Для этого значения тождество формулируется следующим образом:

$1+2+3+...+k+(k+1)=\frac{(k+1)((k+1)+1)}{2}$

Нам необходимо проверить, является ли оно верным. Для этого будем поэтапно преобразовывать его. Для всех слагаемых, кроме последнего, мы уже можем записать представление, которое следует из индукционного предположения, в виде дроби:

$1+2+3+...+k+(k+1)=\frac{k(k+1)}{2}+(k+1)$

Теперь вынесем общий множитель $k+1$:

$\frac{k(k+1)}{2}+(k+1)=(k+1)(\frac{k}{2}+1)$

Приведём вторую скобку к общему знаменателю и запишем всё выражение под общую дробную черту:

$(k+1)(\frac{k}{2}+1)=\frac{(k+1)((k+1)+1)}{2}$

В итоге получаем следующее:

$1+2+3+...+k+(k+1)=\frac{(k+1)((k+1)+1)}{2}$

Это ровно то, что нам нужно было доказать.

***ПРИМЕР 3***

Докажем справедливость следующего тождества:

$1^2+2^2+3^2+...+n^2=\frac{n(n+1)(2n+1)}{6}$

1. Начнём с базы индукции: проверим истинность утверждения для $n=1$:

$1^2=\frac{1(1+1)(2*1+1)}{6}=1$

2. Перейдём к индукционному предположению. Пусть $n+k$ и верно следующее утверждение:

$1^2+2^2+3^2+...+k^2=\frac{k(k+1)(2k+1)}{6}$

Докажем теперь, что утверждение верно для $n=k+1$. То есть мы должны доказать следующее:

$1^2+2^2+3^2+...+k^2+(k+1)^2=\frac{(k+1)((k+1)+1)(2(k+1)+1)}{6}=\frac{(k+1)(k+2)(2k+3)}{6}$

Рассмотрим выражение для $k+1$:

$1^2+2^2+3^2+...+k^2+(k+1)^2$

Все элементы, кроме последнего, можно собрать, согласно нашему индукционному предположению:

$1^2+2^2+3^2+...+k^2+(k+1)^2=\frac{k(k+1)(2k+1)}{6}+(k+1)^2$

Приведём правую часть к общему знаменателю:

$\frac{k(k+1)(2k+1)}{6}+(k+1)^2=\frac{k(k+1)(2k+1)+6(k+1)^2}{6}$

Вынесем $k+1$ как общий множитель:

$\frac{k(k+1)(2k+1)+6(k+1)^2}{6}=\frac{(k+1)(k(2k+1)+6(k+1))}{6}$

Упростим вторую скобку в знаменателе:

$\frac{(k+1)(k(2k+1)+6(k+1))}{6}=\frac{(k+1)(2k^2+7k+6)}{6}$

Разложим вторую скобку на множители:


$\frac{(k+1)(2k^2+7k+6)}{6}=\frac{(k+1)(k+2)(2k+3)}{6}$

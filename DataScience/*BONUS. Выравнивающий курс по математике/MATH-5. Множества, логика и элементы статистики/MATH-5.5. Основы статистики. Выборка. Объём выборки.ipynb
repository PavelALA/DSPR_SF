# <center> MATH-5.5. Основы статистики. Выборка. Объём выборки

***Cтатистику*** можно рассматривать  в более широком смысле — как подраздел математики. В таком понимании статистика включает в себя множество процедур, которые можно использовать для решения какой-либо исследовательской задачи.

***Статистика*** — это набор математических процедур, которые необходимы для организации, обобщения и интерпретации информации.

Что это может быть за информация и откуда она берётся? В идеале мы хотим получить ответы для данных всей генеральной совокупности.

***Генеральная совокупность*** — это совокупность всех объектов, относительно которых предполагается делать выводы при изучении конкретной задачи. То есть, это все те объекты, для которых вы хотите решить задачу.

Конечно, хотелось бы исследовать всю генеральную совокупность. Однако чаще всего это слишком долго, дорого или вовсе невозможно. Например, если мы хотим исследовать мнение всех людей, проживающих в Москве, то часть из них может отказаться участвовать в опросе. По этой причине исследуется не вся генеральная совокупность, а лишь её часть — ***выборка***. То есть из всей генеральной совокупности мы отбираем только часть объектов, проводим на них исследование, а получившиеся результаты обобщаем на всю генеральную совокупность.

***Выборка***, или ***выборочная совокупность*** — это часть генеральной совокупности элементов, которая охватывается экспериментом (наблюдением, опросом).

Однако здесь возникает новая проблема. Допустим, мы хотим опросить всех женщин, чтобы узнать их мнение о новой рекламе косметики, но в выборку возьмём только студенток. Разумеется, в таком случае выводы окажутся неверными, так как мнение женщин более старшего возраста может отличаться. Если мы отберём для выборки только домохозяек, то снова получим ошибочный результат, так как не учтём взгляды работающих женщин.

Здесь возникает понятие ***репрезентативности***, то есть соответствия характеристик выборки характеристикам генеральной совокупности.

Все основные особенности исследуемой группы людей или объектов должны быть одними и теми же у выборки и у генеральной совокупности. Например, если в генеральной совокупности 40 % женщин и 60 % мужчин, такое же соотношение должно сохраниться и в выборке.

***Объём выборки*** — это число случаев, включённых в выборочную совокупность.

Выборки можно условно разделить на ***большие*** и ***малые***, так как в математической статистике используются разные подходы в зависимости от объёма выборки. Считается, что выборки объёмом более 30 можно отнести к большим. Однако это разделение идёт из классической математической статистики — в современном анализе данных количество объектов в выборке обычно исчисляется сотнями тысяч.

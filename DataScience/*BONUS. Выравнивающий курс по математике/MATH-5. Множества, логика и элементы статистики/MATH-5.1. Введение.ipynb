# <center> MATH-5.1 Введение

С одной стороны, для того чтобы сформулировать условия для формирования групп респондентов или наборов изучаемых объектов, вам необходимо научиться работать со множествами и прописывать логические условия. С другой стороны, когда вы справитесь с отбором нужных объектов, вам понадобится кратко и ёмко их описать, проанализировать — здесь на помощь придёт статистика. Таким образом, в этом модуле вы познакомитесь с той областью математики, которая используется при анализе данных.

- научиться реализовывать все основные операции над множествами;

- разобраться с формулированием логических выражений;

- узнать основные понятия из статистики;

- научиться проводить простейший статистический анализ данных.

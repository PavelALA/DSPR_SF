# <center> MATH-1.7. Отрицательные числа и действия над ними. Модуль числа

СВОЙСТВА ОТРИЦАТЕЛЬНЫХ ЧИСЕЛ:

- Любое отрицательное число меньше нуля.
- Любое отрицательное число меньше любого положительного числа.
- При сравнении двух отрицательных чисел больше будет то, которое ближе к нулю (находится правее на числовой прямой).

Также важно уметь реализовывать вычисления с отрицательными числами.

- Если мы вычитаем отрицательное число, это эквивалентно сложению с положительным:

    $5-(-4)=5+4=9$
- Если мы перемножаем два отрицательных числа, в результате получается положительное число

    $(-8)*(-3)=24$
- Если мы делим одно отрицательное число на другое отрицательное число, в результате мы также получаем положительное число:

    $(-24):(-4)=6$

В последних двух примерах фигурируют два отрицательных числа, а в итоге получается положительное. Можно запомнить, что при умножении и делении «минус на минус» даёт «плюс».

На числовой прямой можно увидеть, что есть числа, которые находятся на одинаковом расстоянии от 0, например 4 и -4 или 5 и -5. Чтобы как-то формализовать эту особенность, используется ***модуль***. В англоязычной традиции модуль от числа принято называть ***абсолютным значением***.

***Модуль числа*** — это его расстояние от 0.

Чтобы обозначить модуль, ставят по вертикальной черте с каждой стороны от числа. Например, вместо «модуль от 6» мы можем просто записать:

$|6|$

Модуль  равен:

- x, когда x больше нуля;
- 0, когда x равен 0;
- -x, когда x меньше нуля (то есть модуль «превращает» число обратно в положительное).

Таким образом, когда число положительное или равно нулю, мы оставляем его в первоначальном виде, а когда оно отрицательное, мы меняем его на положительное, добавляя минус перед ним.

СВОЙСТВА МОДУЛЯ:

- $|a|>=0$ - модуль от любого числа всегда является неотрицательным числом.
- $|a|*|b|=|a*b|$ - модуль произведения любых чисел равен произведению модулей этих чисел.
- $|u|=a|$ - это то же самое, что $u=+-a$.

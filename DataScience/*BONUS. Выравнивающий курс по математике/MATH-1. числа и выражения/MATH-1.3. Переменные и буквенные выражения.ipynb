# <center> MATH-1.3. Переменные и буквенные выражения

Буквы, которые содержатся в буквенных выражениях, называются переменными. Переменная — это символ, не имеющий фиксированного значения. Например, в выражении $n^2 + 10n - 5$ ($n^2$ - это n умноженное на n) переменная n может принимать абсолютно любое значение: 1, 2, 3, ...

Если в выражении используется фиксированное число, его называют ***константой***.

***Математическое выражение*** — это обычно одна переменная/константа или комбинация переменных и констант, которые соединены математическими операциями.

Если число и переменная соединены операцией умножения, их называют коэффициентами. Например, $3x^2$, $\frac{-2}{3}$, $\sqrt{5m}$.

Чтобы понять, чему будет равно выражение при заданном значении переменной, необходимо подставить это значение в выражение, упростить его и вычислить результат.

Например, нам необходимо найти значение выражения $5*x^2+2x+7$, если x принимает значение -2.

Итак, ответ равен 23.

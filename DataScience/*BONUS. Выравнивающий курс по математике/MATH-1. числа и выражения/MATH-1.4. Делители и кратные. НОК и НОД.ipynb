# <center> MATH - 1.4. Делители и кратные. НОК и НОД

Возьмём два целых числа и разделим одно на другое:

6/3=2

Мы получили 2 — это целое число. Такое деление называют ***делением нацело***.

Если первое число делится на второе без остатка (то есть нацело), то второе называют делителем первого.

Ещё одно важное понятие, связанное с делением — ***кратное***.

***Кратное для некоторого числа*** — это такое число, которое делится на рассматриваемое число без остатка.



Можно рассматривать кратные не только для каждого числа в отдельности, но и для нескольких чисел. Например, составим список кратных для чисел 3 и 5:

- Кратные для 3 — 3, 6, 9, 12, 15, 18, 21, 24, 27, 30, ...
- Кратные для 5 — 5, 10, 15, 20, 25, 30, …

Мы видим, что есть совпадающие кратные для обоих чисел: 15 и 30. Наименьшее из них равно 15. Поэтому можно сделать вывод, что ***наименьшее общее кратное*** 3 и 5 равно 15. Это можно записать как ***НОК (3, 5) = 15***.

Итак, чтобы вычислить НОК двух чисел A и B методом перечисления, выполните следующие действия:

1. Перечислите несколько первых кратных чисел A и B.
2. Отметьте общие кратные из кратных обоих чисел.
3. Выберите наименьшее общее кратное (НОК) двух чисел.

Однако найти наименьшее общее кратное можно и другим образом. Рассмотрим это на примере чисел 12 и 18.

Разложим каждое из чисел на ***простые множители***:

***Простыми множителями*** называют числа, которые делятся только на себя и на 1. К примеру, 3 является простым множителем, так как делится только на себя и на 1, а 6 — составным, так как делится на себя, а также на 2, 3 и 1.

- 12 = 2*2*3
- 18 = 2*3*3

Теперь выделим все множители, которые встречаются в обоих разложениях.
В нашем случае это 2 и 3. Возьмём их (по одному разу), а также все невыделенные множители и найдём их произведение:

2*3*2*3 = 36

Получаем, что 36 — это наименьшее общее кратное для чисел 12 и 18.

Конечно, наименьшее общее кратное можно искать не только для двух, но и для трёх чисел. Допустим, мы хотим найти общее кратное для 24, 15 и 36.

Выпишем разложение каждого числа на простые множители:

- 24 = 2*2*2*3
- 15 = 3*5
- 36 = 2*2*3*3

Теперь отметим те числа, которые встречаются в каждом разложении. У нас всего одно такое число: 3

Домножим 3 на множители, которые мы не выделили у 24 и 15:

3*2*2*2*5

Из невыделенных множителей числа 36 две двойки мы уже забрали в разложение. Значит, осталось домножить на 3 и вычислить результат:

3*2*2*2*5*3 = 360

Получаем, что ***НОК (15, 24, 36) = 360***.

Кроме наименьшего общего кратного, может пригодиться и ***наибольший общий делитель (НОД)***.

Например, у числа 12 следующие делители: 12, 6, 4, 3, 2, 1. У числа 18 — 18, 9, 6, 3, 2, 1. Общие делители чисел 12 и 18 — 6, 3, 2, 1.

Наибольшее из перечисленных чисел — 6. Значит, 6 является наибольшим общим делителем.

Чтобы найти наибольший общий делитель, необходимо выписать разложение чисел на простые множители и найти множители, которые встречаются в каждом разложении.

Допустим, мы хотим найти наибольший общий делитель для чисел 27 и 45. Выпишем их разложения на простые множители:

- 27 = 3*3*3
- 45 = 3*3*5

В обоих разложениях дважды встречается число 3. Значит, наибольший общий делитель равен 3*3=9. Также это можно записать как НОД (27, 45) = 9.

# <center> MATH-1.6. Пропорции и проценты

***Пропорция*** — это уравнение, которое говорит, что два или более соотношения равны. Например, если в одной упаковке печенья 20 штук, это означает, что две упаковки содержат 40 штук. Таким образом, при увеличении количества упаковок пропорционально увеличивается количество печений.

$\frac{20}{1}=\frac{40}{2}$

Давайте рассмотрим буквенную запись для равенства пропорции. Пропорция читается как «x относится к y так же, как a относится к b».

$\frac{x}{y}=\frac{a}{b}$

Немного преобразуем нашу пропорцию — домножим обе части на y:

$\frac{x}{y}*y=\frac{a}{b}*y$

Сократим x в левой части равенства, а также домножим обе части равенства на b:

$x*b=\frac{a}{b}*y*b$

После упрощения получаем равенство:

$x*b=a*y$

Произведения xb и ay называются ***перекрёстными произведениями***.

Перекрёстные произведения пропорции всегда равны. Если мы хотим проверить, образуют ли два соотношения пропорцию, мы можем просто проверить их перёкрестные произведения. Например, давайте проверим, образуют ли пропорцию дроби 2/16 и 5/40. Реализуем перекрёстные произведения и проверим их равенство:

$\frac{2}{16}=\frac{5}{40}$

$2*40=5*16$

80=80

Итак, мы видим, что 2/16 и 5/40 являются пропорциями, так как их перекрёстные произведения равны.

Пропорции позволяют показать соотношение количества объектов. Например, если у нас есть группа детей, среди которых 20 девочек и 5 мальчиков, то мы можем сказать, что количество девочек относится к количеству мальчиков как 4:1. Также количество мальчиков составляет 20 % от общего количества детей. Вот ещё один способ показать отношение величин — выразить его с помощью процентов.

***Процент*** определяется как заданная часть или количество в каждой сотне. Это дробь со знаменателем 100, которая обозначается символом "%". Например, 5 % от числа — это 5/100.

***Чтобы найти процент от числа, необходимо умножить число на количество процентов и разделить на 100.***

Например, пусть у нас есть 200 товаров и мы знаем, что 15 % из них — бракованные. Мы хотим узнать количество бракованных изделий. В таком случае надо умножить 200 на 15 и разделить на 100:

$200*\frac{15}{100}=30$

Теперь предположим, что мы получили 500 товаров, и нам сообщили, что 30 из них являются бракованными. Как определить, какой процент составляют бракованные товары от общего числа товаров?

***Если мы хотим найти, какой процент одно число составляет от другого, необходимо первое число поделить на второе и умножить результат на 100:***

$\frac{30}{500}*100=6$

В магазине продаётся игровая приставка, которая до распродажи стоила 120$. На распродаже на неё поставили скидку в 25 %. За сколько теперь можно приобрести приставку?

Для начала найдём, чему равняется 25 % от 120 по уже известному нам алгоритму:

$120*\frac{25}{200}=30$

Теперь просто вычтем скидку из изначальной суммы и получим текущую цену: $120-30=90$.

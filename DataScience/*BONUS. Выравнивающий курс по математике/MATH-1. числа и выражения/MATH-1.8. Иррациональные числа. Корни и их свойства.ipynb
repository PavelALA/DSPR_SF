# <center> MATH-1.8. Иррациональные числа. Корни и их свойства

Любое число можно ***возвести в квадрат*** — для этого необходимо умножить число на само себя. Например, если мы хотим возвести 3 в квадрат, нам надо умножить 3 на 3:

$3^2=3*3=9$

У возведения в квадрат есть обратное действие — ***извлечение корня***. Например, мы знаем, что если возвести 4 в квадрат, получится 16. Значит, если извлечь корень из 16, получится 4:

$\sqrt{16}=4$

Говоря про извлечение квадратного корня, мы имеем в виду корень второй степени. На самом деле корень может быть разных степеней. В общем виде корень n-й степени обозначается следующим образом:

$\sqrt[n]{x}$

Например, если мы знаем, что $2^3=8$, то мы можем вычислить $\sqrt[3]{8}=2$.

Конечно, при работе с корнями мы хотим не только извлекать их, но и уметь упрощать более сложные выражения, содержащие корни.

Мы можем «разложить» умножение под знаком корня следующим образом:

$\sqrt[n]{a*b}=\sqrt[n]{a}*\sqrt[n]{b}$

$\sqrt[3]{128}=\sqrt[3]{64*2}=\sqrt[3]{64}*\sqrt[3]{2}=4\sqrt[3]{2}$

Аналогично можно поступить с делением:

$\sqrt[n]{\frac{a}{b}}=\frac{\sqrt[n]{a}}{\sqrt[n]{b}}$

$\sqrt[3]{\frac{1}{64}}=\frac{\sqrt[3]{1}}{\sqrt[3]{64}}=\frac{1}{4}$

***Важно!*** Нет свойств, которые позволяют выполнять следующие действия:

$\sqrt[n]{a+b} != \sqrt[n]{a}+\sqrt[n]{b}$

$\sqrt[n]{a-b} != \sqrt[n]{a}-\sqrt[n]{b}$

$\sqrt[n]{a^n+b^n} != a+b$

Важно отметить, что не из всех чисел можно извлечь корень. Например, из 25 мы можем извлечь целый корень (5), а из 26 — уже нет. Поэтому число $\sqrt{25}$ является ***рациональным***, а число $\sqrt{26}$ — ***иррациональным***.

***Иррациональное число*** — это число, которое не может быть записано в виде простой дроби, в числителе и знаменателе которой будут целые числа.

Далее вы будете встречаться со многими иррациональными числами. Среди них есть такие, для которых существуют отдельные обозначения:

- (Пи)

Люди смогли вычислить число  до более чем квадриллиона знаков после запятой, но в них до сих пор не наблюдается никакой закономерности. Первые несколько цифр выглядят следующим образом:

3.1415926535897932384626433832795…

-  (число Эйлера)

Люди также вычислили e до многих знаков после запятой, и закономерность пока также не удалось обнаружить. Первые несколько цифр выглядят следующим образом:

2.7182818284590452353602874713527…

Рассмотрим ***свойства иррациональных чисел***:

- Иррациональные числа состоят из бесконечных десятичных дробей.
- Результат сложения иррационального и рационального чисел является только иррациональным числом. То есть для иррационального числа x и рационального числа y, x+y- = иррациональное число.
Результат умножения/деления любого иррационального числа на любое ненулевое рациональное число является иррациональным числом. То есть для иррационального числа x и рационального числа y x*y или $\frac{x}{y}$ = иррациональное число.
- Результат сложения, вычитания, умножения и деления двух иррациональных чисел может быть или не быть рациональным числом.

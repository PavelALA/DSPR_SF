# <center> MATH-1.10. Логарифмы и их свойства

***Логарифм*** — это ещё один способ выразить возведение числа в степень.

Мы знаем, что $2^5=32$. Если нас попросят найти число, которое заменяет вопросительный знак в выражении $2^?=32$, то методом проб и ошибок мы сможем выяснить, что это 5.

Но что если мы должны найти число на месте вопросительного знака в равенстве $2^?=30$? Существует ли такое число, что при возведении 2 в степень, равную этому числу, мы получаем 30? Таких чисел мы не знаем. Но как тогда найти ответ? Нужно воспользоваться логарифмами.

Мы можем определить логарифм через возведение в степень. Если мы знаем, что 2 в степени 4 даёт 16, то мы можем сказать, что логарифм по основанию 2 от 16 равняется 4:

$2^4=16 то-же log_216=4$

Давайте обобщим наше понимание логарифма:

$log_b(a)=c$ то же самое что и $b^c=a$

- b — основание логарифма,
- a — показатель логарифма,
- c — значение логарифма.

- Основание логарифма никогда не может быть равно 1. Также оно всегда положительное.
- Показатель логарифма тоже всегда является положительным числом.
- Значение логарифма может быть любым.

Теперь давайте рассмотрим ***правила преобразования и упрощения логарифмических выражений:***

- Логарифм от 1
        - Если мы вычисляем логарифм с показателем, равным 1, то его значение всегда равно 0: $log_a1=0$

- Логарифм с совпадающим основанием и показателем: 
        - Если у логарифма совпадают основание и показатель, то его значение равно 1: $log_aa=1$

- Сложение логарифмов с одинаковым основанием: - Если мы хотим сложить два логарифма, у которых одинаковые основания, можно воспользоваться следующим правилом: $log_ab+log_ac=log_a(b*c)$

- Вычитание логарифмов с одинаковым основанием: $log_ab-log_ac=log_a(\frac{b}{c})$

- Вынесение степени из основания логарифма: $log_{a^p}b=\frac{1}{p}log_ab$

- Вынесение степени из показателя логарифма $log_ab^p=plog_ab$

- Приведение к новому основанию: $log_ab=\frac{log_cb}{log_ca}$

- Основное логарифмическое тождество: $a^{log_ab}=b$

Теперь, когда мы изучили базовые свойства логарифмов, которые помогают преобразовывать логарифмические выражения, давайте разберём ещё пару примеров с преобразованием более сложных выражений.

ПРИМЕР 1

Найти значение выражения $\frac{log_318}{2+log_32}$

В числителе разложим показатель логарифма на два множителя и представим его как сумму логарифмов:

$\frac{log_318}{2+log_32}=\frac{log_3{(9*2)}}{2+log_32}=\frac{log_39+log_32}{2+log_32}=\frac{2+log_32}{2+log_32}=1$

ПРИМЕР 2

Найти значение выражения $log_a{\frac{a^6}{b^4}}$, если $log_ab=-2$.

Сначала разложим логарифм частного как разность логарифмов:

$log_a{\frac{a^6}{b^4}}=log_aa^6-log_ab^4$

Далее вынесем у каждого слагаемого степень показателя как коэффициент:

$log_aa^6-log_ab^4=6log_aa-4log_ab$

Теперь мы можем вычислить первый логарифм по второму свойству (он равен 6*1), а второй логарифм дан в условии. Получаем:

$6log_aa-4log_ab=6*1-4*(-2)=6-(-8)=6+8=14$

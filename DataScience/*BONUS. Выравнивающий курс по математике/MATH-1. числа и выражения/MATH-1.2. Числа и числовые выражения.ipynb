# <center> MATH-1.2. Числа и числовые выражения

По сути, ***число*** — это некоторая идея в нашем сознании, которую мы используем для формализации какого-то количества, длины, веса: например, три игрушки, шесть тарелок, две книги, сто рублей, пять метров, десять килограммов. Каждый день мы обращаемся к числам, порой даже не замечая этого.

Для записи чисел используются ***цифры*** — их десять: 0, 1, 2, 3, 4, 5, 6, 7, 8 и 9.

Цифры относятся к числам, как буквы — к словам. Из букв мы можем составить слово, и оно будет отражать наше представление о чём-то. Так же и из цифр можно составить число, которое будет являться формализацией какого-то представления.

С числами можно производить различные операции. Все возможные операции можно разделить на две группы:

- операции действия,
- операции сравнения.


Операции действия:

- сложение (+);
- вычитание (-);
- умножение (*);
- деление (/).

***Сложение*** — это прибавление к числу одного или более чисел, чтобы получить новое общее число. Например, мы складываем 5 булочек и 6 булочек и получаем 11 булочек.

Числа, которые складываются, называют ***слагаемыми***, а результат сложения называют ***суммой***.

Допустим, мы хотим сложить числа 7 и 9. Мы можем записать эту операцию и её результат следующим образом:

***7+9=16***

***Вычитание*** — это отнимание одного числа от другого.

Число, из которого вычитают другое, называют ***уменьшаемым***, число, которое вычитают, — ***вычитаемым***, а результат вычитания — ***разностью***.

Например, мы хотим вычесть из числа 40 число 15. Можем записать это так:
***40-15=25***


***Произведение*** — это увеличение какого-то числа в определённое число раз.

Например, если мы умножаем 6 на 7, это значит, что мы увеличиваем 6 в 7 раз.

Число, которое увеличивают, называется ***множимым***. Число, которое показывает во сколько раз нужно увеличить множимое, называется ***множителем***. Число, которое получается в результате — ***произведение***.

Умножим 7 на 8 — получаем:

***7 * 8=56***

***Деление*** — это уменьшение числа в какое-то число раз.

Например, если мы делим 100 на 10, это значит, что мы хотим уменьшить 100 в 10 раз.

Число, которое мы уменьшаем, называется ***делимым***. ***Делитель*** — число, которое показывает, во сколько раз мы уменьшаем делимое. ***Частное*** — полученный результат.

Например, разделим число 10 на 2:

***10 / 2 = 5***



***Однако в математике обычно выполняют не отдельные действия, а составляют целые выражения.***

***Выражение*** — это любое сочетание из чисел, букв и операций.

Выражения бывают двух видов:

- числовые,
- буквенные.

Числовые выражения состоят из чисел и знаков математических операций. Например, следующие выражения являются числовыми:

***3+5+9***

***25 + 6 * 3***



Числовые выражения могут быть и более сложными, например:

***(5 + 6 + 3) * 4 + 5 * (6 + 34 : 2)***

***Порядок операций*** — это правило, согласно которому сначала реализуют вычисления в скобках, далее выполняют деление или умножение (слева направо, в зависимости от того, что наступит раньше), а сложение или вычитание производят в последнюю очередь (слева направо, в зависимости от того, что наступит раньше).

Таким образом, порядок операций всегда следующий:

- Скобки (), {}, [].
- Деление и умножение.
- Сложение и вычитание.

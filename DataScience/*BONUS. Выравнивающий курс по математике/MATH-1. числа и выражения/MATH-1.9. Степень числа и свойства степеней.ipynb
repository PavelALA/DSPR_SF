# <center> MATH-1.9. Степень числа и свойства степеней

Формула степени с a в качестве ***основания*** и n в качестве ***показателя*** степени приведена ниже:

$a^n=a*a*a....*a - n_{раз}$

По сути, для того чтобы возвести число в степень, необходимо умножить его на само себя n раз.

Существуют ***свойства степеней***, которые помогают упрощать степенные выражения:

1. Возведение числа в нулевую степень.

Любое число при возведении в нулевую степень превращается в единицу:

$a^0=1$

2. Возведение числа в степень 1.

Любое число при возведении в степень 1 остаётся неизменным:

$a^1=a$

3. Произведение степеней с одинаковым основанием:

$a^m+a^n=a^{m+n}$

4. Частное степеней с одинаковым основанием:

$\frac{a^m}{a^n}=a^m-n$

5. Отрицательная степень:
$\frac{1}{a^m}=a^{-m}$

6. Степень в степени:

$(a^m)^n=a^{m*n}$

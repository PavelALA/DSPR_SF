В первом модуле курса мы поговорим о вычислениях: начнём с понятия числа, вспомним правила математических вычислений и углубимся в более сложные сущности — пропорции, отрицательные и иррациональные числа, степени и логарифмы.

***ЦЕЛИ ДАННОГО МОДУЛЯ:***

- рассмотреть понятие числ и правила математических вычислений;

- научиться реализовывать буквенные выражения и переводить задачи на язык математических моделей;

- познакомиться с понятиями «делитель» и «кратное», научиться находить НОК и НОД;

- изучить пропорции и проценты и научиться решать задачи с использованием этих сущностей;

- узнать свойства иррациональных чисел и научиться с ними работать;

- разобраться со степенями и логарифмами, их свойствами и преобразованиями степенных и логарифмических выражений;

- познакомиться с математическими обозначениями и символами, понимание которых упростит чтение математических записей.

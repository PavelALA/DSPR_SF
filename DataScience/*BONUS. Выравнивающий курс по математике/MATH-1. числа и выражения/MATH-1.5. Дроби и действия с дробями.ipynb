# <center> MATH-1.5. Дроби и действия с дробями

Для начала вспомним, что такое дробь. Представим, что у нас есть пирог, который мы разделили на 4 части. Одну часть мы съели (закрасим её красным).

Так как мы забрали одну часть из четырёх, мы можем записать эту долю в виде дроби следующим образом:

$\frac{1}{4}$

Число 1 над дробной линией называется ***числителем***: оно показывает, сколько частей мы отбираем (в нашем случае закрашиваем красным). Число 4 под чертой называется ***знаменателем***: оно показывает, на сколько равных долей делится целое.

Теперь давайте поставим рядом несколько пирогов. Первый разделён на четыре доли, и две из них закрашены красным цветом, но, как мы видим, это составляет половину пирога. Второй пирог разделён на шесть долей, и три из них закрашены красным — это вновь половина пирога. Наконец, третий пирог разделён на две половины, и одна из них закрашена красным.

Поскольку во всех случаях закрашена половина, можно сделать вывод, что представленные дроби равны: 2/4 = 3/6 = 1/2.

*Также из этого наблюдения можно вывести следующее свойство: если умножить или разделить числитель и знаменатель на одно и то же число, дробь останется такой же (за исключением случая деления на ноль).*

При взгляде на два куска пирога мы обычно можем определить, какой из них больше. Аналогично с дробями — существует простой способ сравнить их друг с другом.

Скажем, нам нужно сравнить 1/3 и 2/7. Если бы у них были одинаковые знаменатели, сравнить было бы легко: больше была бы дробь с большим числителем. Однако здесь знаменатели отличаются. Для того чтобы дроби было проще сравнивать, нам необходимо привести их к общему знаменателю.

***Один из методов нахождения общего знаменателя двух или более дробей — это умножение знаменателей друг на друга: . Также в качестве общего знаменателя можно брать наименьшее общее кратное. Однако в нашем случае НОК для 3 и 7 также будет равен 21.***

Теперь, когда мы нашли общий знаменатель, необходимо заменить знаменатель каждой дроби на общий.

Ранее мы упомянули, что можно домножить и числитель, и знаменатель дроби на одно и то же число. Воспользуемся этим. Домножим в дроби  и числитель, и знаменатель на 7:

$\frac{1}{3}=\frac{7}{21}$

Таким же образом приведём к новому знаменателю дробь 2/7:

$\frac{2}{7} = \frac{6}{21}$

Теперь, когда у этих дробей одинаковый знаменатель, мы можем сравнить их. Семь долей больше, чем шесть, поэтому 7/21 больше, чем 6/21. Математически мы можем записать это так:

$\frac{7}{21}>\frac{6}{21}$


- ***Сложение дробей***

Чтобы сложить дроби с одинаковым знаменателем, надо просто сложить их числители, а знаменатель оставить прежним:

$\frac{3}{13}+\frac{5}{13}=\frac{3+5}{13}=\frac{8}{13}$

Если у дробей разные знаменатели, дроби сначала необходимо привести к общему знаменателю, а затем сложить их числители:

$\frac{4}{15}+\frac{3}{5}=\frac{4}{15}+\frac{9}{15}=\frac{4+9}{15}=\frac{13}{15}$

- ***Вычитание дробей***

Вычитание дробей происходит по тому же принципу, что и сложение. Из числителя одной дроби вычитается числитель другой, а знаменатель остаётся прежним:

$\frac{3}{4}-\frac{2}{4}=\frac{3-2}{4}=\frac{1}{4}$

В случае с разными знаменателями всё происходит аналогично, но первоначально, разумеется, необходимо привести дроби к общему знаменателю:

$\frac{10}{15}-\frac{3}{5}=\frac{10}{15}-\frac{9}{15}=\frac{10-9}{15}=\frac{1}{15}$

- ***Умножение дробей***

При умножении дробей отдельно перемножаются числители и отдельно — знаменатели.

***Важно***, что для умножения приводить дроби к общему знаменателю не требуется (это не повлияет на результат, но усложнит вычисления):

$\frac{3}{5}*\frac{4}{7}=\frac{3*4}{5*7}=\frac{12}{35}$

- ***Деление дробей***

Чтобы разделить одну дробь на другую, необходимо перевернуть вторую дробь (поменять местами числитель и знаменатель), а затем просто умножить первую на перевёрнутую вторую:

$\frac{3}{5}:\frac{4}{7}=\frac{3}{5}*\frac{7}{4}=\frac{3*7}{5*4}=\frac{21}{20}$

В последнем примере получилась дробь, у которой числитель больше знаменателя. Такая дробь называется ***неправильной***.

В этом случае мы можем выделить у дроби целую часть. Так как в число 21 целиком один раз «умещается» число 20, то целая часть будет равна 1 и останется ещё одна часть:

$\frac{21}{20}=\frac{20}{20}+\frac{1}{20}=1\frac{1}{20}$

Если числитель и знаменатель можно разделить на одно и то же число, дробь можно сократить. Например, для дроби 10/15 мы можем и числитель, и знаменатель разделить на 5:

$\frac{10}{15}=\frac{2}{3}$

Кроме изученных ранее типов дробей, существует ещё один — десятичные дроби. Это дроби, у которых знаменатель кратен 10 (равен 10, 100, 1000 и т. д.) и которые записываются определённым образом. К примеру, $\frac{3}{4}$ можно записать в десятичном виде как 0.75. Если мы захотим представить в десятичном виде дробь $\frac{7}{25}$, то можем преобразовать её в дробь со знаменателем, кратным 10 (для этого домножим числитель и знаменатель дроби на 4), и получить $\frac{28}{100}$, что можно записать как 0.28. Количество знаков после запятой равняется количеству нулей в знаменателе числа.

Рассмотрим ещё один пример. Пусть у нас есть дробь $\frac{1}{3}$. Такую дробь мы не сможем перевести в десятичный вид, поскольку нет такого числа, которое при умножении на 3 дало бы какую-то степень числа 10. В таком случае получается бесконечная дробь — 0.333333…

Алгоритм округления следующий:

1. Узнаём, до какого знака после точки/запятой необходимо округлить дробь.
2. Смотрим, какая цифра идёт после интересующей нас цифры:
    - Если эта цифра равна 5 и более, округляем в большую сторону.
    - Если эта цифра меньше 5, оставляем всё как есть.

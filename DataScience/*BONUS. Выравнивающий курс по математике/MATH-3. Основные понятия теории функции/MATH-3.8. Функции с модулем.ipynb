# <center> MATH-3.8. Функции с модулем

Общий вид функции с модулем:

$y=a|x-h|+k$

Коэффициент $a$ показывает, насколько график растягивается по вертикали, а также его направление (вверх — при положительном $a$, вниз — при отрицательном $a$). Константы $h$ и $k$ показывают, насколько график сдвигается по горизонтали и вертикали.

Давайте рассмотрим самую простую функцию с модулем: $y=|x|$. Коэффициент $a$ здесь равен 1, а константы равны 0. Таким образом, график никак не растягивается и не сжимается, а также не сдвигается. Он будет выглядеть так:

Давайте рассмотрим шаги, которые необходимо предпринять при построении графиков функций с модулем

Допустим, мы хотим построить график следующей функции:

$f(x)=|x-1|+5$

- $a=1$, поэтому график направлен вверх с коэффициентом наклона 1.
- $h=1$, $k=5$, поэтому вершина графика смещена на 1 единицу вправо и на 5 единиц вверх от начала координат.

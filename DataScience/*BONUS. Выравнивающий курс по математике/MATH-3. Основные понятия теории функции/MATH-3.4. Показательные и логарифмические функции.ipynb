# <center> MATH-3.4. Показательные и логарифмические функции

## <center> ПОКАЗАТЕЛЬНАЯ ФУНКЦИЯ

***Показательная функция*** — это функция вида $f(x)=a*b^x$, где a и b — вещественные числа, причём b — положительное число.

С помощью показательной функции оценивают различные процессы, например биологические. Если мы знаем, что популяция кроликов увеличивается в 1.5 раза в месяц и что изначально у нас есть 100 кроликов, то мы можем узнать, сколько кроликов будет через год именно с помощью показательной функции:

$100*1.5^12=12975$

Вид графика показательной функции зависит от значения основания степени. Если основание степени больше 1, то функция возрастает и её график имеет следующий вид:

Если основание степени изменяется от 0 до 1, то функция убывает и её график выглядит так:

![](https://lms.skillfactory.ru/assets/courseware/v1/7b2082e8392c917f98e0cfdb907fe093/asset-v1:SkillFactory+DSPR-2.0+14JULY2021+type@asset+block/MFTIDSLIGHT_MATH_3_4_4.png)

***Основание степени не может быть меньше нуля, как и не может быть равно 0 или 1.***

Показательная функция обладает ***следующими свойствами***:

- ***Область определения*** — все вещественные числа, т. е. $D(f):(-\infty;+\infty)$.
- ***Область значений*** — все положительные числа, так как положительное число в какой-то степени никогда не станет отрицательным числом или нулём: $E(f):(0;+\infty)$.
- Не является чётной или нечётной.



## <center> ЛОГАРИФМИЧЕСКАЯ ФУНКЦИЯ

С помощью логарифмических функций описывают математические модели различных явлений: раковины моллюсков, рога горных козлов, галактика Млечный Путь — для описания закрученности этих объектов используют как раз логарифмы.

К примеру, ниже представлена иллюстрация логарифмической закрученности галактики:

![](https://lms.skillfactory.ru/assets/courseware/v1/15afe7bc2977a0f94d33768e4a214b35/asset-v1:SkillFactory+DSPR-2.0+14JULY2021+type@asset+block/MFTIDSLIGHT_MATH_3_4_5.png)

Функция, заданная формулой $y=log_2x$, $(a>0, a\neq1)$ , называется ***логарифмической функцией с основанием a***.

Примеры логарифмических функций:
- $y=log_3x$
- $y=log_0.1(x-3)$

Логарифмическая функция обладает следующими ***свойствами***:

- Область определения — все положительные числа, т. е. $D(f):(0;+\infty)$.
- Область значений — все вещественные числа , т. е. $E(f):(-\infty;+\infty)$.
- Не является чётной или нечётной.
- При основании логарифма $0<a<1$ функция убывает на $D(f):(0;+\infty)$, а при $a>1$ — возрастает на $(0;+\infty)$.

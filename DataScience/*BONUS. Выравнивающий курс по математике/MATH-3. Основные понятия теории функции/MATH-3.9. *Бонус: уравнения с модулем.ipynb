# <center> MATH-3.9. *Бонус: уравнения с модулем

Есть несколько способов решения таких уравнений, и мы рассмотрим два основных: решение уравнений по определению и перебор промежутков.



## <center> РЕШЕНИЕ УРАВНЕНИЙ ПО ОПРЕДЕЛЕНИЮ

Решим уравнение $|x+4|=1$.

Равенство будет выполняться в двух случаях:

- $x+4=1$
- $x+4=-1$

Решим уравнение из первого случая:

$x+4=1$

$x=1-4$

$x=-3$

Теперь — из второго:

$x+4=-1$

$x=-1-4$

$x=-5$

Получаем два ответа: -3 и -5.

Рассмотрим более сложный пример — решим уравнение $|2-5x^2|=3.

Как и в предыдущем случае, равенство распадается на два уравнения:

$2-5x^2=3$

$2-5x^2=-3$

По очереди решим оба уравнения:

$2-5x^2=3$

$2-5x^2-3=0$

$-5x^2-1=0$

$-5x^2=1$

$x^2=-\frac{1}{5}$

Таким образом, нам нужно число, которое после возведения в квадрат даёт отрицательный результат. Таких чисел не существует, поэтому решений нет.

$2-5x^2=-3$

$-5x^2=-5

$x^2=1$

$x=\pm1$

Получаем два корня:

$x_1=1$

$x_2=-1$


## <center> ПЕРЕБОР ПРОМЕЖУТКОВ

Пусть у нас есть уравнение $|2-x|=5-4x$

На промежутке $x\leq2$ под модулем будет неотрицательное значение, поэтому можно просто снять модуль — получаем:

$2-x=5-4x$

$4x-x=5-2$

$3x=3$

$x=1$

Мы рассматривали промежуток $x\leq2$ и получили корень 1, который принадлежат этому промежутку. Значит, это действительно корень.

Теперь рассмотрим второй промежуток — $x>2$. На нём значение под модулем уже получается отрицательным, поэтому необходимо добавить минус:

$-(2-x)=5-4x$

$-2+x=5-4x$

$x+4x=5+2$

$5x=7$

$x=1.4$

Число 1.4 не входит в промежуток, который мы рассматриваем, значит, его нельзя считать корнем. В итоге получаем один корень: $x=1$.

Промежутков может быть и больше двух. Решим уравнение $|x-1|+|x-2|=3$

Промежутков может быть и больше двух. Решим уравнение 

$|x-1|=\begin{cases}x-1, если х-1\geq0\\ -(x-1), если x-1<0\end{cases}\iff |x-1|=\begin{cases}x-1, x\geq1\\ -(x-1), если x<1\end{cases}$

Второй модуль, разумеется, можно открыть аналогично:

$|x-2|=\begin{cases}x-2, если x-2\geq0\\ -(x-2), если x-2<0\end{cases}\iff|x-2|=\begin{cases}x-2, если x\geq2\\ -(x-2), если x<2\end{cases}$

В итоге получаем три ситуации:

$|x-1|+|x-2|=3\iff\begin{cases}x-1+x-2=3, x\geq2\\ x-1-x+2=3, 1\leq x<2\\ -x+1-x+2=3, x\leq1\end{cases}$

Итак, мы получили три уравнения, каждое из которых решается в определённом числовом промежутке. Решим по очереди все три уравнения:

1. $x-1+x-2=3$

$2x=6$

$x=3$

Мы решали это уравнение для всех x, которые не меньше 2. Корень входит в рассматриваемый промежуток, значит, он будет являться ответом.

2. $x-1-x+2=3$

$1=3$

Получаем противоречие, поэтому делаем вывод, что для этого случая корней нет.

3. $-x+1-x+2=3$

$-2x=3-3$

$-2x=0$

$x=0$

Мы решали это уравнение для всех значений x, которые меньше 1. Число 0 входит в этот промежуток.

В итоге получаем, что у уравнения два корня: 0 и 3.

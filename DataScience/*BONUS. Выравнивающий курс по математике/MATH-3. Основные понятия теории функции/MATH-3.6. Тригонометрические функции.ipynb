# <center> MATH-3.6. Тригонометрические функции

***Единичная окружность*** — это окружность с радиусом 1 (единица) и с центром в начале координатной плоскости.

Если взять любой угол и одну из его сторон «положить» на ось , то вторая сторона угла пересечёт тригонометрическую окружность в какой-то точке.

![](https://lms.skillfactory.ru/assets/courseware/v1/e4f960e0a6fc309ba7db48b29b5f9756/asset-v1:SkillFactory+DSPR-2.0+14JULY2021+type@asset+block/MFTIDSLIGHT_MATH_3_6_2.png)

Разумеется, у получившейся точки пересечения будет координата по оси абсцисс (значение косинуса) и координата по оси ординат (значение синуса). Координаты могут быть как положительными, так и отрицательными в зависимости от того, в какой четверти угол пересечёт окружность:

![](https://lms.skillfactory.ru/assets/courseware/v1/6d386563537bcf17e047c8b01de07697/asset-v1:SkillFactory+DSPR-2.0+14JULY2021+type@asset+block/MFTIDSLIGHT_MATH_3_6_3.png)

к как радиус окружности — единица, то синус и косинус не могут быть больше 1 и меньше -1:

- $-1\leq sin\alpha\leq1$
- $-1\leq cos\alpha\leq1$

Теперь дадим определения оставшимся тригонометрическим функциям — ***тангенсу*** и ***котангенсу***. Тангенс и котангенс произвольного числа вводятся через отношения уже известных нам синуса и косинуса.

Тангенсом числа $\alpha$ называется отношение синуса числа $\alpha$ к косинусу числа $\alpha$:

- $tg \alpha=\frac{sin \alpha}{cos \alpha}$


Котангенсом числа $\alpha$ называется отношение косинуса числа $\alpha$ к синусу числа $\alpha$:

- $ctg \alpha=\frac{cos \alpha}{sin \alpha}$

Под градусами для каждого угла также есть запись с  $\pi$— это представление величины угла в радианах, ещё одной мере, в которой можно измерять углы. Выразить величину угла в радианах — значит, выразить её через $\pi$, равное 180 градусам. К примеру, если угол равен 90 градусов, это то же самое, что $\frac{\pi}{2}$, а $\frac{3\pi}{4}$ — это 135 градусов.

Несмотря на то что в окружности всего 360 градусов, мы можем находить значения тригонометрических функций для углов больше 360, например для 710 или 1000 градусов. То есть в качестве аргумента может выступать абсолютно любое вещественное число. И, конечно, это можно отобразить графически.

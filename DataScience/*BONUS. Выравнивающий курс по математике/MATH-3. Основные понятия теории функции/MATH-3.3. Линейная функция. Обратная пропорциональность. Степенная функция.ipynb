# <center> MATH-3.3. Линейная функция. Обратная пропорциональность. Степенная функция

## <center> ЛИНЕЙНАЯ ФУНКЦИЯ

***Линейной*** называют функцию, которую можно представить в виде следующей формулы:

$y=kx+b$

где x — независимая переменная, k и b — некоторые числа.

Графиком линейной функции является **прямая**. Чтобы построить её, необходимо найти любые две точки, которые принадлежат графику, и провести через них прямую.

Допустим, мы хотим построить график функции $y=3x+2$. Можно взять два значения x, подставить их в функцию и таким образом найти две точки для графика:

- если x=-1, то y=-1.
- если x=1, то y=5.

Мы можем отметить эти точки на координатной плоскости, а затем провести через них прямую:

![график линейной функции](https://lms.skillfactory.ru/assets/courseware/v1/310d83539d7f85b3aa194bfef5e9d478/asset-v1:SkillFactory+DSPR-2.0+14JULY2021+type@asset+block/MFTIDSLIGHT_MATH_3_3_1.png)

Обратите внимание, что график функции пересёк ось ординат (ось y) в точке 2. Эта ордината совпадает с числом b из формулы y=kx+b. В линейных функциях так происходит всегда, поэтому, чтобы упростить себе задачу, одну из точек можно, не задумываясь, отмечать на оси y: она будет равна числу b.

Если число b будет равно 0, то линейная функция будет иметь вид y=kx и одной из её точек всегда будет точка 0 на оси ординат. Эта точка — начало координат. Такую функцию, полученную из функции y=kx, называют ***прямой пропорциональностью***. Графиком прямой пропорциональности является прямая, проходящая через начало координат.

## <center> ОБРАТНАЯ ПРОПОРЦИОНАЛЬНОСТЬ

***Обратной пропорциональностью*** называют зависимость, при которой увеличение независимой переменной ведёт к уменьшению значения функции.

К примеру, если у нас есть функция $y=\frac{1}{x}$, то мы можем отметить, что чем больше будет x, тем меньше будет y.

В общем виде функция обратной пропорциональности определяется следующим образом:

$y=\frac{k}{x}$

В качестве графика функции выступают две ветви ***гиперболы***.

***Гипербола*** — это кривая с двумя ветвями, т. е. кривая, состоящая из двух частей определённой формы.

![график гиперболы](https://lms.skillfactory.ru/assets/courseware/v1/f5ad8e7d772759a48d403bbe51e52e88/asset-v1:SkillFactory+DSPR-2.0+14JULY2021+type@asset+block/MFTIDSLIGHT_MATH_3_3_4.png)

Если k положительное, то ветви располагаются в верхней правой и нижней левой четвертях, если k отрицательное — в верхней левой и нижней правой:

![График гипербол](https://lms.skillfactory.ru/assets/courseware/v1/9e8fae2b1a4d6cea6d47bd386e202e4e/asset-v1:SkillFactory+DSPR-2.0+14JULY2021+type@asset+block/MFTIDSLIGHT_MATH_3_3_5.png)

Таблица со значениями для этой функции может выглядеть так:
- x= 1,2,3,6,-1,-2,-3,-6
- y= 6,3,2,1,-6,-3,-2,-1

## <center> СТЕПЕННАЯ ФУНКЦИЯ

В своей простейшей форме ***степенные функции*** описывают ситуации, когда зависимая переменная прямо пропорциональна степени независимой переменной.

Степенная функция имеет следующий вид:

$f(x)=k*x^n$,

где k и n — вещественные числа.

В зависимости от чётности степени независимой переменной график будет представлять собой:

- ***гиперболу*** — для нечётной степени (слева);
- ***параболу*** — для чётной степени (справа).

Исключение будет составлять лишь случай, при котором n=1: в этой ситуации функция превращается в линейную. Близость к оси y у любого графика зависит от величины коэффициента n: чем больше коэффициент, тем быстрее у функции увеличивается значение и тем ближе она к оси y.

![](https://lms.skillfactory.ru/assets/courseware/v1/05a93cfc204f9bfc3d4f4d1e97a00251/asset-v1:SkillFactory+DSPR-2.0+14JULY2021+type@asset+block/MFTIDSLIGHT_MATH_3_3_8.png)

Если степень является нечётным числом, то функция постоянно возрастает, а если чётным — убывает на промежутке $(-\infty;0)$ и возрастает на промежутке $(0;+\infty)$.

Разумеется, как и все предыдущие функции, степенную функцию можно построить по точкам. Построим самую известную из степенных функций — квадратичную. Пусть уравнение функции будет таким:

$y=2x^2$

Можно составить следующую таблицу значений аргументов и значений функций:
- x:-1,1,0,-2,2,-3,3,-4,4
- y=2,2,0,8,8,18,18,32,32

График будет выглядеть так:

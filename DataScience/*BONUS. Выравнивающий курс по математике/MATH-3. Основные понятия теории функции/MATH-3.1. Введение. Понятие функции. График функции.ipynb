# <center> MATH-3.1. Введение. Понятие функции. График функции

ЦЕЛИ МОДУЛЯ:

- разобраться с понятиями «функция» и «график функции»;

- научиться исследовать функции и анализировать их свойства;

- узнать основные виды функций:

    - линейную;
    - степенную;
    - показательную;
    - логарифмическую;
    - тригонометрические.


***Функция (функциональная зависимость)*** — это такая зависимость одной переменной от другой, где каждому значению независимой переменной соответствует только одно значение зависимой.

В нашем примере с квартирой её площадь — это независимая переменная, а цена — зависимая. Разумеется, каждой площади соответствует только одна цена.

Независимую переменную иначе называют ***аргументом***, а зависимую — ***функцией*** от этого ***аргумента***.

$P=S*2000000+3000000$

где P — цена квартиры, а S — её площадь.

Такое равенство называют формулой.

***Формула*** — это своего рода закон, по которому мы находим одно значение, зная другое. Например, зная, как связаны цена квартиры и её площадь, мы можем подставлять в формулу различные значения одной величины и получать другую.

Итак, мы научились находить значения функции, которые соответствуют разным значениям аргумента, то есть мы можем получить много пар, состоящих из значения аргумента и значения функции. Теперь мы можем отметить точки, соответствующие этим парам, на координатной плоскости. Значение аргумента будем отмечать на горизонтальной оси (абсцисса / ось абсцисс), а значение функции — на вертикальной (ордината / ось ординат). Если отметить все возможные точки, мы получим график функции

***График функции*** — это такое множество точек (x;y), где x — аргумент, а y — значение функции, которое соответствует данному аргументу.

Разумеется, невозможно найти все существующие точки, поэтому будем брать несколько и рисовать примерный эскиз графика, ориентируясь на них. Чем больше пар «значение аргумента — значение функции» мы найдём, тем больше будет точек и точнее график.

Допустим, необходимо построить график функции, заданной формулой $y=\frac{6}{x+3}$ где $-2\leq x\leq+3$

Мы можем подставлять в функцию любые значения аргумента и находить соответствующие им значения функции. Например, возьмём целые значения. Оформим результаты в виде таблицы:

x=-2;-2;0;1;2

y=6;3;2;1.5;1.2;1

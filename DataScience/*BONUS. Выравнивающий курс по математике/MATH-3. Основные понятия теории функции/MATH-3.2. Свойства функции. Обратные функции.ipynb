## <center> ОБЛАСТЬ ОПРЕДЕЛЕНИЯ

По сути, ***область определения*** — это множество всех значений аргументов, к которым мы можем применять рассматриваемую функциональную зависимость, и с точки зрения математики это будет «разрешено».

Пусть у нас есть функция $f(x)=\frac{1}{x}$.

Можно использовать в качестве аргумента любое число, кроме нуля, так как на ноль делить нельзя. Это значит, что областью определения функции являются абсолютно все вещественные числа, за исключением нуля.

Более строго, математически это утверждение можно записать следующим образом:

$D(f(x)):x\in(-\infty;0)\cup(0;+\infty)$

$D(f(x))$ здесь читается как «область определения функции f от x». Утверждение целиком можно прочитать как «Область определения функции f от x:  x принадлежит интервалу от минус бесконечности до нуля в объединении с интервалом от нуля до плюс бесконечности».

***ПРИМЕР 1***

Найдём область определения функции $\frac{x^2}{(x-3)(x-5)}$.

Как видим, в числителе можно использовать любое значение , а в знаменателе есть два значения, при которых возникнет деление на 0.

Итак, областью определения данной функции будут все вещественные числа, за исключением 3 и 5.

Формально мы можем записать это следующим образом:

$D(f(x)):x\in(-\infty;3)\cup(3;5)\cup(3;+\infty)$

***ПРИМЕР 2***

Найдём область определения для немного более сложной функции: $f(x)=\frac{\sqrt{x+2}}{x^2-9}$



В данном примере необходимо обратить внимание как на числитель, так и на знаменатель. В числителе находится квадратный корень от x+2. Корень мы можем извлекать только из неотрицательных чисел, поэтому x должен быть больше или равен -2. Выражение в знаменателе принимает нулевое значение при x=3 и при x=-3.

Таким образом, получаем следующую область определения:

$D(f(x)):x\in[-2;3)\cup(3;+\infty)$

Мы рассмотрели ограничения для знаменателя дроби и квадратного корня. Также существуют ограничения для логарифмов: ***в основании логарифма не может быть неположительного числа или единицы, а показателем логарифма должно быть строго положительное число***:


$log_ab:\begin{cases}
            a>0\\
            a\neq1\\
            b>0
            \end{cases}$

Ограничения на показатель логарифма наглядно видно на графике логарифмической функции:

## <center> ОБЛАСТЬ ЗНАЧЕНИЙ

Разумеется, не все функции могут принимать какое угодно значение. Например, функция $y=x^2$ может принимать только неотрицательные значения, так как никакое число в квадрате никогда не может быть меньше нуля. Это очень хорошо видно на графике: все точки находятся в верхней (положительной) половине координатной плоскости.

![гипербола](https://lms.skillfactory.ru/assets/courseware/v1/f13700b984171c4dc0a967c2f1721729/asset-v1:SkillFactory+DSPR-2.0+14JULY2021+type@asset+block/MFTIDSLIGHT_MATH_3_2_3.png)

Можно записать это формально следующим образом:

$E(f(x)):f(x)\in[0;+\infty)$

Рассмотрим пример — найдём область значений функции:

$f(x)=\frac{15}{x^2+1}$

Мы видим, что числитель здесь фиксированный (равен константе, а значит, не изменяется), а знаменатель равен 1 при x=0. Если x принимает любое другое значение, то знаменатель будет больше единицы. Причём чем больше x по модулю, тем больше значение $x^2+1$. Если x примет бесконечно большое значение, то значение дроби будет очень маленьким, бесконечно близким к нулю. Наибольшее из возможных значений дроби будет равно 15 (если знаменатель равен 1). Таким образом, получаем следующую область значений:

$E(f(x)):f(x)\in(0:15]$

## <center> ТОЧКИ ПЕРЕСЕЧЕНИЯ С ОСЯМИ АБСЦИСС И ОРДИНАТ

Следующий важный шаг в исследовании функции — поиск точек, в которых функция пересекает ось абсцисс (горизонтальная) и ось ординат (вертикальная).

Давайте посмотрим на график ниже. Можно отметить, что в точках, где график пересекает ось x, координата по y равна нулю. И наоборот: в точках пересечения с осью ординат координаты по x принимают значение 0.

![График пересечения](https://lms.skillfactory.ru/assets/courseware/v1/3b332a94be6668d0ca3d3094cc563a7a/asset-v1:SkillFactory+DSPR-2.0+14JULY2021+type@asset+block/MFTIDSLIGHT_MATH_3_2_4.png)

Эта удобная особенность позволяет без проблем находить пересечения с осями.

Например, пусть у нас есть функция $f(x)=\frac{x^2-2x}{x-1}$

Для начала найдём точки пересечения с осью ординат. Для этого просто подставим 0 вместо x: $f(x)=\frac{0^2-2*0}{0-1}=0$

Таким образом, пересечение с осью ординат находится в точке (0;0).

Теперь найдём пересечение с осью x. Для этого заменим значение функции на ноль:

$0=\frac{x^2-2x}{x-1}$

Чтобы значение дроби было равно 0, необходимо, чтобы числитель был равен 0:

$(x^2-2x)=0$

Это квадратное уравнение, у которого без труда можно найти два корня:

$x_1=0$
$x_2=2$

Получаем, что точки пересечения с осью абсцисс — (0;0) и (2;0).


##  <center> ЧЁТНОСТЬ И НЕЧЁТНОСТЬ

Функция является ***чётной***, если для всех значений x выполняется равенство $f(x)=f(-x)$. Это означает, что функция одинакова на положительной и отрицательной полуплоскостях для оси x и симметрична относительно оси y.

***Полуплоскость*** — это множество точек плоскости, лежащих по одну сторону от некоторой прямой на этой плоскости.

Например, функция $y=x^2$ является чётной:

![четная функция](https://lms.skillfactory.ru/assets/courseware/v1/288c2437c987a0a459dcff1bc8b1d102/asset-v1:SkillFactory+DSPR-2.0+14JULY2021+type@asset+block/MFTIDSLIGHT_MATH_3_2_5.png)

Ещё один пример чётной функции — $f(x)=|x|$, поскольку функция модуля принимает одинаковое значение для x и -x.

Функция является ***нечётной***, если $f(-x)=-f(x)$ для всех x. График нечётной функции будет симметричен относительно начала координат.

Например, функция $f(x)=x^3$ является нечётной:

![нечетная функция](https://lms.skillfactory.ru/assets/courseware/v1/45a4031b6bbfbcea0609167291f167eb/asset-v1:SkillFactory+DSPR-2.0+14JULY2021+type@asset+block/MFTIDSLIGHT_MATH_3_2_6.png)



Давайте попробуем исследовать несколько функций на чётность и нечётность.

***ПРИМЕР 1***

Дана функция $5*x^2-4|x|$.

Подставим -x вместо x — получим:

$5*(-x)^2-4|-x|$

По правилам математических преобразований, это выражение равняется исходному:

$5*x^2-4|x|=5*(-x)^2-4|-x|$

Получается, что функция является чётной.

***ПРИМЕР 2***

Дана функция $f(x)=x-4x^3$.

Подставим -x вместо x — получим:

$f(-x)=-x-4(-x)^3=-(x-4x^3)$

То есть мы получили равенство $f(-x)=-f(x)$. Значит, функция является нечётной. Это можно увидеть и на графике: он симметричен относительно начала координат.

![пример 2](https://lms.skillfactory.ru/assets/courseware/v1/a4f8d4a6265bc34bc1030f23e890633b/asset-v1:SkillFactory+DSPR-2.0+14JULY2021+type@asset+block/MFTIDSLIGHT_MATH_3_2_7.png)

ПРИМЕР 3

Дана функция $f(x)=x^2-x$.

Подставим -x вместо x — получим:

$f(-x)=-x^2-(-x)=x^2+x

Данный результат не удовлетворяет ни условиям чётности, ни условиям нечётности. Отсутствие симметрии можно также наблюдать на графике функции:

## <center> ВОЗРАСТАНИЕ И УБЫВАНИЕ

Функция называется ***возрастающей***, если при увеличении её аргумента увеличивается и значение функции.

Формально можно записать это следующим образом:

$x_1>x_2$ и $f(x_1)>f(x_2)$

С ***убывающей*** функцией всё наоборот: её значение уменьшается с увеличением значения аргумента.

Формальная запись:

$x_1>x_2$ и $f(x_1)<f(x_2)$

## <center> ОБРАТНАЯ ФУНКЦИЯ

***Обратной*** называют функцию, которая обращает зависимость, выражаемую рассматриваемой функцией. Например, если функция делает из числа a число b, то обратная ей функция делает из числа b число a.

Обратные функции бывают полезны, когда известна зависимость первой величины от второй, но нужно узнать, как вторая величина зависит от первой. Например, нам известна площадь для ремонта и, исходя из неё, мы вычисляем необходимые объёмы строительных материалов. Обратная функция позволила бы определить, какую площадь можно отремонтировать с помощью заранее известного объёма строительных материалов.

Давайте рассмотрим, как найти обратную функцию, на примере. Пусть у нас есть функция $y=6x-4$.

Выразим из этого равенства x:

$y=6x-4$

$-6x=-y-4$

$6x=y+4$

$x=\frac{y+4}{6}$

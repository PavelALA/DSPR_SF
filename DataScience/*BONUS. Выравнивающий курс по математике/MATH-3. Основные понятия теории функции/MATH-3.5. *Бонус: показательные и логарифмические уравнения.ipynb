## <center> ПОКАЗАТЕЛЬНЫЕ УРАВНЕНИЯ

***Показательным*** называется уравнение, содержащее неизвестную величину в показателе степени.

Рассмотрим ***три основных метода решения показательных уравнений***.



### ПРИВЕДЕНИЕ К ОДНОМУ ОСНОВАНИЮ

Этот метод предполагает, что обе части уравнения (левую и правую) необходимо представить как степени с одинаковым основанием. После этого можно «отбросить» основания степени, так как если две показательных функции равны и у них равны основания степеней, должны быть равны и показатели степеней.

Пусть у нас есть следующее уравнение:

$4^{x-1}=2^x*\sqrt2$

$\sqrt2$ можно записать как 2 в степени $\frac{1}{2}$:

$4^{x-1}=2^x*2^{\frac{1}{2}}$

Число 4 мы можем представить как 2 в квадрате:

$(2^{x-1})^2=2^x*2^{\frac{1}{2}}$

Вспомним свойства преобразования степеней и применим их к правой и левой частям уравнения:

$2^{2x-2}==2^{x+\frac{1}{2}}$

Мы видим, что слева и справа у степеней одинаковые основания. Значит, чтобы выражения были равны, показатели степеней тоже должны быть одинаковыми:

$2x-2=x+\frac{1}{2}$

$2x-x=2+\frac{1}{2}$

$x=2+0.2$

#x=2.5$

### ВЫНЕСЕНИЕ ЗА СКОБКУ

Ещё один популярный метод — разложение одной из частей уравнения на множители.

Например, попробуем решить следующее уравнение:

$3^x+3^{x+3}=84$

Вынесем слева за скобку $3^x$:

$3^x(1+3^3)=84$

$3^x(1+27)=84$

$3^x*28=84$

Поделим обе части на 28:

$3^x=3$

Теперь мы, по сути, свели уравнение к предыдущему способу (приведение к одному основанию) — применим его:

$3^x=3^1$

$x=1$

### ЗАМЕНА ПЕРЕМЕННОЙ

$36^x-9*6^x+8=0$

Обозначим $6^x$ за t — получаем:

$t^2-9t+8=0$

Решаем квадратное уравнение и находим два корня:

$t_1=1; t_2=8$

В результате обратной замены получаем, что $6^x=1$ или $6^x=8$.

Для первого случая легко найти корень — $x=0$.

Со вторым случаем уже сложнее, но здесь мы можем вспомнить одно из свойств логарифмов и преобразовать правую часть уравнения следующим образом:

$6^x=8$

$6^x=6^{log_68}$

Отбрасываем основания степеней и получаем ответ:

$x=log_68$

## <center> ЛОГАРИФМИЧЕСКИЕ УРАВНЕНИЯ

***Логарифмическим*** называется уравнение, в котором неизвестная содержится в показателе логарифма.

Для решения логарифмических уравнений используют два основных метода: ***метод потенцирования и метод замены переменной***.

***Потенцирование*** — действие, обратное логарифмированию.

### МЕТОД ПОТЕНЦИРОВАНИЯ

Метод потенцирования предполагает, что в левой и правой части уравнения появляются логарифмы с одинаковым основанием. Тогда можно «отбросить» логарифмы и просто приравнять их показатели.

Например, мы хотим решить следующее уравнение:

$log_3(x^2-3x+3)=log_3(x^2+x+3)$

Удобно, что слева и справа уже стоят логарифмы. Однако прежде чем приступить к решению, необходимо прописать область определения: т. е. изучить, при всех ли значениях х показатели логарифма будут положительными:

$\begin{cases}
    x^2-3x+3>0\\
    x^2+x+3>0
    \end{cases}$

Если построить графики для обеих этих функций (например, по точкам — это будут параболы), то мы увидим, что они всегда находятся в положительной полуплоскости. Значит, значения этих выражений всегда положительные вне зависимости от значения x.

Приравниваем показатели логарифмов:

$x^2-3x+3=x^2+x+3$

$x^2-x^2-3x-x+3-3=0

$-4x=0$

$x=0$

К сожалению, не всегда сразу с обеих сторон уравнения находятся логарифмы с одинаковыми основаниями. Рассмотрим пример:

$log_5(x-3)=2$

Обозначим область определения: $x>3$.

Теперь преобразуем правую часть в логарифм с основанием 5 — получаем:

$log_5(x-3)=log_5(25)$

После этого можно оставить только показатели логарифма и приравнять их:

$x-3=25$

$x=25+3$

$x=28$



### МЕТОД ЗАМЕНЫ ПЕРЕМЕННОЙ

Пусть у нас есть уравнение:

$log_2^2x-4log_2x+4=0$

Мы видим, что здесь повторяется $log_2x$ — заменим это выражение на t:

$t^2-4t+4=0$

У этого уравнения будет только один корень t=2.

Делаем обратную замену:

$log_2x=2$

$log_2x=log_24$

$x=4$

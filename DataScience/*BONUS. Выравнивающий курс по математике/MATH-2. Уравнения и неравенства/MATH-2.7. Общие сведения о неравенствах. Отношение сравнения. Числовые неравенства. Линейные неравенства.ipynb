# <center> MATH-2.7. Общие сведения о неравенствах. Отношение сравнения. Числовые неравенства. Линейные неравенства

В жизни нам зачастую приходится сравнивать величины: больше — меньше, много — мало. В математике мы также часто встречаемся со сравнением чисел. Для этого используются следующие знаки:

-  > — больше,
-  < — меньше.

Неравенство a>b означает, что a-b — положительное число.

Неравенство a<b означает, что a-b — отрицательное число.

Перед тем как приступить к решению различных неравенств, давайте рассмотрим их свойства и примеры использования.

- Если a>b и b>c, то a>c.	
- Если, a>b то a+c>b+c.	
- Если a>b и m>0, то am>bm.	
- Если a>b и m<0, то am<bm.	
- Если a>b и c>d, то a+c>b+d.	
- Если a,b,c,d>0 и a>b, а c>d, то ac>bd.	
- Если  a>=0, b>=0 и a>b, то $a_n>b^n$, где n принадлежит N.	
- Если a и b — числа одного знака и не равны нулю, а также если a>b, то $\frac{1}{a}<\frac{1}{b}$.

## <center> ЛИНЕЙНЫЕ НЕРАВЕНСТВА

Линейным неравенством называется неравенство вида axVb, где вместо символа v стоит один из знаков <,>, <=,>= где a=!0, а b — любое число.

Ответом любого неравенства является ***промежуток***. Промежуток включает в себя два числа либо число и знак бесконечности. Запись промежутка начинается с числа, которое расположено левее, и заканчивается числом, расположенным правее на числовой оси.

# <center> MATH-2.1. Введение. Общие сведения об уравнениях

Уравнения и неравенства можно встретить в любой области математики. Также они прочно связаны с Data Science и другими компьютерными науками. Например, если мы хотим предсказать цену квартиры по её характеристикам, необходимо найти определённые параметры — сделать это можно, решив систему уравнений.

Ранее вы уже сталкивались с ***числовыми равенствами***. Например, оно может выглядеть следующим образом:

34-10=16+8

Здесь левая и правая части равны — в обоих случаях получится 24.

Иногда в равенстве могут быть неизвестные. Например, если бы мы не знали, что должно быть на месте  -10, то равенство выглядело бы следующим образом:

34+x=16+8

В этом случае мы имеем дело с ***уравнением***.

***Уравнение*** — это равенство с неизвестным числом.

***Решить уравнение*** — значит найти такие значения переменных, которые делают уравнение верным. Эти значения называют корнями уравнения.

Например, корнем уравнения 5x+3=13 будет являться число 2, так как если подставить 2 вместо , равенство будет верным.

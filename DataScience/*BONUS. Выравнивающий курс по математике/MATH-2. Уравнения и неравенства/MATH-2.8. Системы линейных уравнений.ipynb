# <center> MATH-2.8. Системы линейных уравнений

***Система уравнений*** — это условие, состоящее в одновременном выполнении нескольких уравнений, входящих в систему.

Система двух уравнений обозначается следующим образом: 

$\begin{cases}
    p(x;y)=0,\\
    q(x;y)=0
  \end{cases}
$

Здесь $p(x;y)=0$  и $q(x;y)=0$ — это какие-то уравнения, в которых есть два неизвестных, x и y.

***Решением*** этой системы называется пара чисел $(x;y)$, которая одновременно является решением и первого, и второго уравнения, т. е. при подстановке обращает их оба в верные равенства.

Если все уравнения, входящие в систему, являются линейными, то представляют собой уравнения вида $ax+by+c=0$, то система называется ***линейной***. Если хотя бы одно из уравнений системы не является линейным, то система ***называется нелинейной***.

Далее мы изучим два метода, которые используются для решения систем уравнений: ***метод подстановки*** и ***метод сложения***. Мы будем рассматривать их применение только на линейных системах, однако важно отметить, что и на нелинейных эти методы можно применять, просто это может быть сложнее.

## МЕТОД ПОДСТАНОВКИ

Суть метода заключается в том, что в одном уравнении нужно выразить одну переменную через другую и подставить её во второе уравнение. Таким образом получится одно уравнение с одной неизвестной.

Давайте рассмотрим пример применения такого метода. Пусть у нас есть следующая система уравнений:

$\begin{cases}
        x=5y\\
        2x-7y=6
    \end{cases}$

В первой строке переменная x уже выражена через y. Подставляем это выражение во вторую строку:

$2*5y-7y=6$

$10y-7y=6$

$3y=6$

$y=2$

После того как мы нашли y, необходимо найти значение x. Для этого подставляем y в ту часть, где мы выразили x через y:

$x=5y=5*2=10$

В итоге получаем ответы: x=10, y=2. Иногда такую пару записывают в скобках: (10;2).

## МЕТОД СЛОЖЕНИЯ

Суть метода заключается в том, что мы так складываем уравнения системы (левую часть одного уравнения — с левой частью другого, правую часть одного — с правой частью другого), что одна из переменных сокращается. Иногда для этого требуется предварительно умножить одно или оба уравнения на некоторые коэффициенты.

Например, мы хотим решить следующую систему:

$\begin{cases}
        4y+5x=-10\\
        3y-5x=17
    \end{cases}$

Можно заметить, что в первом уравнении есть слагаемое 5x, а во втором — -5x. Это значит, что уравнения можно сложить. При этом переменная х сократится и останется одно линейное уравнение с одной переменной:

4y+5x=-10

3y-5x=17

7y+0=7

7y=7-0

7y=7

y=1

Подставляем полученное значение в первое уравнение:

4*1+5x=-10

5x=-10-4

5x=-14

$x=-\frac{14}{5}$

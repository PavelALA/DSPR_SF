# <center> MATH-2.2. Линейные уравнения и их применение

***Линейное уравнение*** — это уравнение с переменной (переменными) первой степени и одной или несколькими константами (числами, которые не изменяются).

Под переменной первой степени мы понимаем переменную вида x,y,z и так далее. Если переменная возводится в степень, отличную от 1, то она уже не является переменной первой степени. Например, $x^2$ — это переменная второй степени, а $y^3$ — переменная третьей степени.

Общий вид такого уравнения: ax+b=0 , где:

- a и b — какие-то вещественные числа;
- x — переменная.

Наша цель при решении линейных уравнений с одной переменной — найти значение переменной, которое делает уравнение верным.

Таким образом, алгоритм решения линейных уравнений можно сформулировать следующим образом:

1. Раскрыть скобки, если они есть.
2. Перенести слагаемые с переменной в одну часть (обычно в левую), а все остальные (т. е. константы) — в другую часть (обычно в правую).
3. Привести подобные слагаемые.
4. Разделить правую и левую части уравнения на коэффициент перед неизвестным.

## <center> ПРИМЕНЕНИЕ ЛИНЕЙНЫХ УРАВНЕНИЙ


В машину Daewoo Matiz может поместиться небольшая семья. Однако при желании в машину могут влезть ещё два человека.

Из скольких людей состоит небольшая семья, если известно, что при желании в Daewoo Matiz помещается пять человек?

В этой задаче нам неизвестен размер небольшой семьи, поэтому именно его мы возьмём за x. Получаем, что если к x добавить  2(ещё двух человек), то получится  5(максимальная вместимость машины):

x+2=5

x=5-2=3

# <center> MATH-2.5. Квадратные уравнения и их применение

Квадратным уравнением называется уравнение вида $ax^2+bx+c$, где a, b и c — любые действительные числа (также a не может быть равен 0), причём коэффициенты a, b, c имеют свои названия:

- a — старший коэффициент,
- b — второй коэффициент,
- c — свободный коэффициент.

Характерной чертой квадратных уравнений является переменная в степени 2, причём это должна быть самая  большая степень в уравнении.

Примеры квадратных уравнений:

- $2x^2+3x+4=0$
- $-4x^2+6x+9=0$
- $3(x^2+8)-4x=9x+10$

Квадратное уравнение называется ***приведённым***, если его старший коэффициент равен 1.

Пример:
- $x^2-45x+86+0$



## <center> АЛГОРИТМ РЕШЕНИЯ КВАДРАТНЫХ УРАВНЕНИЙ:

1. Найти коэффициенты.

Помните:

- коэффициент a всегда стоит перед $x^2$;
- коэффициент b всегда стоит перед x;
- коэффициент c — свободный коэффициент, просто число.

2. Найти дискриминант.

***Дискриминантом*** называется вспомогательный аргумент, который входит в формулу для поиска корней уравнения. Он вычисляется по следующей формуле:

$D=b^2-4ac$

Зная значение дискриминанта, можно понять, сколько корней имеет уравнение:

- если дискриминант положительный — два корня;
- если дискриминант равен 0 — один корень;
- если дискриминант меньше нуля — решений нет.

2. Найти корни.

Корни квадратного уравнения вычисляются по формулам:

$x_1=\frac{-b+\sqrt{D}}{2a}$

$x_2=\frac{-b-\sqrt{D}}{2a}$

Если дискриминант равен 0, то значения этих двух выражений, разумеется, совпадут.

Рассмотрим применение этого алгоритма на примере. Пусть у нас есть следующее квадратное уравнение:

$4x^2+3x-7=0$

1. Определим коэффициенты:

a=4,b=3,c=-7

2. Вычислим дискриминант:

$d=3^2-4*4*(-7)=121*$

3. Найдём корни:

$x_1=\frac{-3+\sqrt{121}}{2*4}=\frac{8}{8}=1$

$x_1=\frac{-3-\sqrt{121}}{2*4}=-\frac{14}{8}=-\frac{7}{4}=-1\frac{3}{4}$

Ещё один способ решения квадратных уравнений — с использованием ***теоремы Виета***.

Суть ***теоремы Виета*** заключается в том, что для уравнения $ax^2+bx+c$ с корнями $x_1$ и $x_2$ выполняются следующие равенства:

$x_1+x_2=\frac{-b}{a}$

$x_2*x_2=\frac{c}{a}$

В частности, для приведённого квадратного уравнения $x^2+px+q=0$ сумма корней равна $-p$, а произведение равно $q$.

# <center> MATH-2.6. Уравнения высших порядков

Уравнения высших порядков — это уравнения, в которых наибольшая степень у переменной превышает 2.

Не все такие уравнения можно решить самостоятельно (без использования компьютера). Однако для нахождения корней тех уравнений, что поддаются решению, существуют специальные методы — давайте рассмотрим их.



## <center> РАЗЛОЖЕНИЕ НА МНОЖИТЕЛИ

В самых простых случаях многочлен удаётся разложить на множители, удачно группируя друг с другом слагаемые.

К примеру, мы хотим решить следующее уравнение:


$2x^3-3x^2-8x+12=0$

Мы можем разложить левую часть этого уравнения на множители методом группировки:

$2x^3-3x^2-8x+12=0=x^2(2x-3)-4(2x-3)=(x^2-4)(2x-3)$

Также можно разложить первое слагаемое на множители, используя формулу сокращённого умножения:

$(x^2-4)(2x-3)=(x-2)(x+2)(2x-3)=0$

Итак, у нас есть три множителя, произведение которых равно 0. Следовательно, либо $(x-2)=0$, либо $(x+2)=0$, либо $(2x-3)=0$.

Получаем три корня: $x_1=2, x_2=-2, x_3=1.5$

## <center> ПОДБОР КОРНЯ

Рассмотрим его на примере: решим уравнение $x^3-9x^2+26x-24=0$

К сожалению, здесь не получится простого разложения на множители, поэтому попробуем подобрать корень. Если корень целый, то он обязательно должен быть делителем свободного коэффициента, т. е. числа -24. Если мы будем перебирать все делители, то в итоге придём к выводу, что корнем является $x=2$.

Это значит, что мы можем так разложить многочлен на множители, что одна из скобок будет $x-2$. Таким образом, нам просто нужно последовательно находить слагаемые, у которых можно будет извлечь общий множитель, чтобы в скобках осталось $x-2$. Получаем:

$x^3-2x^2-7x^2+14x+12x-24=x^2(x-2)-7x(x-2)+12(x-2)$

Выносим $(x-2)$ как общий множитель — в итоге получается:

$(x-2)(x^2-7x+12)=0$

Из первой скобки получаем корень $x=2$, из второй скобки получаются корни 3 и 4.

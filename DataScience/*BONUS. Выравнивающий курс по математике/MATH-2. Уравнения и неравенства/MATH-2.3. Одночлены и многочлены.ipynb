# <center> MATH-2.3. Одночлены и многочлены

***Одночлен*** — это такая математическая запись, которая представляет собой произведение чисел и переменных, возведённых в степени с натуральным показателем. Также одночленами являются все числа и переменные.

Примеры одночленов: $5xy, 6ab^2, \frac{3}{8}yx, 7, -y^3$

- Число, стоящее перед одночленом, называют ***числовым коэффициентом***.
- Произведение переменных с их степенями в одночлене называют ***буквенной частью***.

Прежде чем использовать одночлен, его приводят к стандартному виду.

***Стандартным видом одночлена*** называют такой его вид, при котором одночлен не содержит повторяющихся переменных и скобок и не имеет в своей записи более одного числа.

Например, одночлены $4x*x*y$ и $7*x*4$ представлены не в стандартном виде, так как у первого есть повторяющаяся переменная, а второй содержит более одного числа. Если привести эти одночлены в стандартный вид, то получится $4x^2y$ и $28x$.

Одночлены редко встречаются поодиночке — чаще они появляются в составе многочленов.

***Многочлен*** — это математическое выражение, состоящее из переменных, коэффициентов, операций сложения, вычитания и умножения, а также  неотрицательных целых степеней. Такие выражения также называются ***полиномами***.

Примеры многочленов: 
- $x-5$
- $2x^2-16x+7x+9$
- $2a^3b^2-3b^2+2a+1$

Многочлены используются практически в каждой области математики для отражения взаимосвязи разных величин и показателей. Полиномы также являются «строительными блоками» для других типов математических выражений, таких как рациональные выражения (дробей, у которых и в числителе и в знаменателе стоят выражения с переменными).

***Выражение, не являющееся многочленом***
1. $2x+x^{\frac{1}{2}}$
2. $\frac{x}{y}+2y$
3. $6x^{-2}+2x-3$

***Почему это не многочлен***
1. В многочлене не может быть показательных выражений (степеней, выраженных переменными) и нецелых степеней.
2. В многочленах могут быть дроби, однако в знаменателе не может быть переменных.
3. В многочленах не может быть отрицательных степеней.



Прежде чем работать с многочленом, его, как и одночлен, принято приводить в стандартный вид.

Рассмотрим многочлен $3xy*5y-2y*7x^2-x*4yx$. Он записан в не очень удобном формате, и эту запись можно упростить.

Для этого надо:

1. Привести каждый многочлен к стандартному виду.
2. Сложить подобные одночлены (если они есть).

Итак, для начала приведём все одночлены, которые входят в многочлен, к стандартному виду:

$3xy5y-2*y*7x^2-4x^2y=15xy^2-14x^2y-4x^2y$

Теперь приведём подобные по аналогии с тем, как мы делали это ранее в буквенных выражениях. Здесь подобными являются второе и третье слагаемое, так как у них одинаковый набор переменных в одинаковых степенях. Получаем:

$15xy^2-14x^2y-4x^2y=15xy^2-18x^2y$

Кроме того, бывает необходимо узнать ***степень многочлена***. Чтобы вычислить её, надо привести многочлен к стандартному виду и найти максимальную из степеней входящих в него одночленов.

Предположим, нам нужно найти степень многочлена $2x3yx+12z^2=6x^2y=12z^2$

Степень первого члена равна 3 (у переменной x степень 2, у переменной y — степень 1, в сумме — 3), а степень второго — 2. Значит, степень многочлена — 3.

## <center> ДЕЙСТВИЯ С МНОГОЧЛЕНАМИ

Самые простые операции — это сложение и вычитание. При сложении и вычитании мы просто записываем многочлены друг за другом, соединяя их плюсами и минусами и приводим подобные.

Пусть мы хотим сложить многочлены $xy^2+5xyz+3z$ и $7x-2xyz+3xy^2$

Записываем их сумму:

$xy^2+5xyz+3z+7x-2xyz+3xy^2$

Приводим подобные:

$xy^2+5xyz+3z+7x-2xyz+3xy^2=4xy^2+3xyz+3z+7x$

Умножение реализуется чуть сложнее: для этого надо умножить первый многочлен на каждый элемент второго, а затем привести подобные слагаемые.

Допустим, мы хотим перемножить многочлены $xy+x$ и $y+1$. Получаем:

$(xy+x)(y+1)=xy*y+x*y+xy*1+x*1=xy^2+xy+xy+x+xy^2+2xy+x$

Рассмотрим ещё один, более сложный, пример. Будем умножать многочлен $ab+a^2-b^2+3$ на многочлен $b+3a$.

$(ab+a^2-b^2+3)(b+3a)=ab^2+a^2b-b^3+3b+3a^2b+3a^3-ab^2+9a$

Приводим подобные слагаемые:

$ab^2+a^2b-b^3+3b+3a^2b+3a^3-ab^2+9a=4a^b-4ab^2=3a^3-b^3+9a+3b$

Иногда бывает нужно не перемножить многочлены, а наоборот, получить из результата их умножения изначальные выражения. Это может пригодиться для упрощения сложных математических выражений, с которыми мы столкнёмся в дальнейшем. Такую операцию называют ***разложением многочлена на множители***.



## <center> ВЫНЕСЕНИЕ ОБЩЕГО МНОЖИТЕЛЯ

Алгоритм вынесения общего множителя:

1. Найти такое число или переменную, на которые делится каждый член многочлена.
2. Разделить каждый член многочлена на эту переменную (или число), записать многочлен в скобках.
3. Саму переменную (число) записать перед получившимся в скобках многочленом.
4. Повторять шаги 1-3, пока переменные или числа, на которые делится многочлен, не закончатся.

Рассмотрим пример $3a^3b^2+6ab^5-12a^2b^3$.

Видим, что можно вынести 3:

$3a^3b^2+6ab^5-12a^2b^3=3(a^3b^2+2ab^5+4a^2b^3)$

Также выносим a:

$3(a^3b^2+2ab^5+4a^2b^3)=3a(a^2b^2+ab^5+4ab^3)$

И, наконец, выносим $b^2$:

$3a(a^2b^a+ab^5+4ab^3)=3ab^2(a^2+ab^3+4ab)$

# <center> MATH-1.9. Системы линейных неравенств

Несколько неравенств с одной переменной образуют ***систему неравенств***, если требуется найти все общие решения заданных неравенств.

***Решить систему неравенств*** — значит найти те значения переменной, которые являются решениями каждого из неравенств.

Разберём алгоритм решения системы неравенств на примере системы с двумя линейными неравенствами.

$\begin{cases}
    2x-1>3\\
    3x-1<11
    \end{cases}$

Сначала решим первое неравенство:

$2x-1>3$

$2x>4$

$x>2$

Теперь решим второе неравенство:

$3x-1<11$

$3x<12$

$x<4$

В нашем случае штриховки совпали на интервале от 2 до 4. Следовательно, ответ будет записываться следующим образом:

$x\in(2:4)$

Рассмотрим ещё один пример решения системы неравенств.

$\begin{cases}
    5(x+1)\leq3(x+3)+1\\
    \frac{2x-1}{7}\leq\frac{x+1}{2}
    \end{cases}$

Откроем скобки в первом неравенстве, а второе умножим на 14:

$\begin{cases}
    5x+5\leq3x+10\\
    {2(2x-1)}\leq7(x+1)
    \end{cases}$

$\begin{cases}
    5x+5\leq3x+10\\
    4x-2\leq7x+7
    \end{cases}$

Перенесём неизвестные влево, а числа — вправо. Приведём подобные слагаемые и разделим на коэффициенты перед переменной:

$\begin{cases}
    2x\leq5\\
    -3x\leq9
    \end{cases}$


$\begin{cases}
    x\leq2.5\\
    x\geq-3
    \end{cases}$

Получается, что x должен быть одновременно больше или равен -3 и меньше или равен 2.5. Следовательно, получаем ответ: $x\in[-3;2.5]$

# <center> MATH-4.9. Основы теории векторов. Векторы в пространстве и на плоскости и действия над ними

***Вектор*** — это направленный отрезок.

Вектор обозначается «стрелкой»: запись $\overrightarrow{AB}$ означает, что точка  A — начало, а точка B — конец вектора AB.

![](https://lms.skillfactory.ru/assets/courseware/v1/9f3c4c3d5b1035aef1069c3393e7f024/asset-v1:SkillFactory+DSPR-2.0+14JULY2021+type@asset+block/MFTIDSLIGHT_MATH_4_9_1.png)

Для работы с векторами нам потребуется умение находить их ***координаты***. Чтобы найти координаты вектора, необходимо из координат конца вектора вычесть соответствующие координаты его начала.

Другими словами, если известны координаты двух точек $A(x_1,y_1,z_1)$ и $B(x_2,y_2,z_2)$, то координаты вектора  вычисляются с помощью простого покоординатного вычитания конца вектора из его начала:

$\overrightarrow{AB}=(x_2-x_1;y_2-y_1;z_2-z_1)$

Кроме координат, у вектора также есть ***длина***.

***Длиной вектора*** называют длину отрезка, то есть расстояние от начала вектора до его конца.

Если известны координаты вектора $\overrightarrow{AB}(x;y;z)$, то его длина вычисляется по формуле:

$\overrightarrow{AB}=\sqrt{x^2+y^2+z^2}$

Длина вектора обозначается двумя вертикальными чертами: |a| или $|\overrightarrow{AB}|$.

Если длина вектора равна 1, его называют ***единичным***, а если 0 — ***нулевым***. У нулевого вектора начало совпадает с концом.

## <center> СУММА И ПРОИЗВЕДЕНИЕ ВЕКТОРОВ

Складываются векторы достаточно просто — покоординатно:

$(x_1;y_1;z_1)+(x_2;y_2;z_2)=(x_1+x_2;y_1+y_2;x_1+z_2)$

Скалярным произведением векторов называется произведение длин этих векторов на косинус угла между ними:

$(\overrightarrow{a},\overrightarrow{b})=|\overrightarrow{a}|*|\overrightarrow{b}|*\cos(\overrightarrow{a},\overrightarrow{b})$

Чтобы находить скалярное произведение векторов, совсем не обязательно знать их длины и угол — достаточно знать координаты, так как можно просто сложить покоординатные произведения:

$(\overrightarrow{a},\overrightarrow{b})=\overrightarrow{a}*\overrightarrow{b}=x_1x_2+y_1y_2+z_1z_2$

Интересно, что, используя данные формулы, можно получить ещё одну, с помощью которой можно найти косинус угла между векторами, зная их координаты:

$\cos(\overrightarrow{a},\overrightarrow{b})=\frac{(\overrightarrow{a},\overrightarrow{b})}{|\overrightarrow{a}|*|\overrightarrow{b}|}=\frac{x_1x_2+y_1y_2+z_1z_2}{\sqrt{x_1^2+y_1^2+z_1^2}*\sqrt{x_2^2+y_2^2+z_2^2}}$

# <center> MATH-4.8. Основные фигуры стереометрии. Многогранники: призма и пирамида

***Многогранник*** — это поверхность, составленная из многоугольников и ограничивающая некоторое геометрическое тело. По сути, многогранники — это объёмные многоугольники.


## <center> ПРИЗМА

***Призма*** — это многогранник, две грани которого являются равными многоугольниками, лежащими в параллельных плоскостях, а остальные грани — параллелограммами, имеющими общие стороны с этими многоугольниками. Иными словами, это многогранник, в основаниях которого лежат равные многоугольники, а боковые грани являются параллелограммами.

***Правильная призма*** — это прямая призма, в основании которой лежат правильные многоугольники.

Для многогранников нам необходимо уметь находить площадь поверхности и объём. Как правило, площадь для оснований и боковой поверхности вычисляют отдельно. Если необходима полная площадь фигуры, эти значения складывают.

Чтобы найти суммарную площадь оснований призмы, необходимо просто найти площадь одного основания $S_{осн}$ и умножить её на 2. Формула для поиска площади основания зависит от того, какая фигура лежит в основании призмы.

Чтобы найти площадь боковой поверхности, необходимо умножить периметр основания на высоту пирамиды:

$S_{бок}=h*P_{осн}$

Тогда площадь полной поверхности равна:

$S_{общ}=2*S_{осн}+S_{осн}$

Чтобы найти объём призмы, необходимо умножить площадь основания на высоту:

$V=S_{осн}*h$

Давайте рассмотрим треугольную призму, у которой в основании лежит треугольник со сторонами 13, 14 и 15, а боковые стороны призмы перпендикулярны основанию (т. е. являются высотами) и равны 16:

![](https://lms.skillfactory.ru/assets/courseware/v1/bb971fd0115701425625461fc80b1389/asset-v1:SkillFactory+DSPR-2.0+14JULY2021+type@asset+block/MFTIDSLIGHT_MATH_4_8_4.png)

Найдём площадь поверхности этой призмы, а также её объем.

Площадь основания нам уже известна: мы искали площадь для треугольника с такими сторонами в одном из предыдущих юнитов — она равна 84.

Периметр основания равен 13+14+15=42

Тогда площадь боковой поверхности равна:  $S_{бок}=42*16=672$

Площадь общей поверхности: $S_{общ}=2*84+672=840$

Вычислим объём призмы: $V=84*16=1344$

## <center> ПИРАМИДА

***Пирамида*** — это многогранник, одна из граней которого — произвольный многоугольник (основание), а остальные грани — треугольники с общей вершиной (боковые грани).

Пирамида называется ***правильной***, если в её основании лежит правильный (многоугольник, у которого все стороны равны) многоугольник, а отрезок, соединяющий вершину пирамиды с центром её основания, является её высотой.

Площадью полной поверхности пирамиды называется сумма площадей всех её граней, площадью боковой поверхности пирамиды — сумма площадей её боковых граней. Вот как это выглядит для пирамиды:

$S_{полн}=S_{бок}+S_{осн}$

Чтобы найти объём пирамиды, необходимо воспользоваться следующей формулой:

$V=\frac{1}{3}*S_{осн}*h$

Предположим, мы хотим найти площадь всей поверхности, а также объём правильной четырёхугольной пирамиды, в основании которой лежит квадрат:

![](https://lms.skillfactory.ru/assets/courseware/v1/23e4e1e89cf0a89a141663cc05d3fceb/asset-v1:SkillFactory+DSPR-2.0+14JULY2021+type@asset+block/MFTIDSLIGHT_MATH_4_8_9.png)

Известно, что сторона квадрата равна 8, высота пирамиды равна 6.

Сначала мы можем найти объём. Площадь основания найти легко: $S=8^2=64$. Подставляем площадь в формулу объёма: $V=\frac{1}{3}*64*6=128$.

С площадью полной поверхности сложнее. Площадь основания мы уже знаем, а площадь боковой поверхности сразу найти не получится. Поскольку боковыми гранями пирамиды являются треугольники, необходимо найти площадь одного такого треугольника. Таким образом, вся боковая поверхность пирамиды будет в четыре раза больше.

Так как AOB — равнобедренный прямоугольный треугольник, мы можем найти длину OB:

$AO^2+BO^2=64$

$2BO^2=64$

$BO^2=32$

$BO=\sqrt{32}$

Теперь по теореме Пифагора из треугольника SOB мы можем найти длину SB:

$SB^2=SO^2+OB^2$

$SB^2=6^2+32=36+32=68$

$SB=\sqrt{68}$

В треугольнике ASB известны все стороны. Площадь можно найти, например, с помощью формулы Герона.

Полупериметр будет равен: $p=\frac{\sqrt{68}+\sqrt{68}+8}{2}=\sqrt{68}+4$

Площадь треугольника:

$S=\sqrt{4*4*53}=8\sqrt{13}$

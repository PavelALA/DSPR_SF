# <center> MATH-4.4. Основы тригонометрии (синус, косинус и тангенс). Теорема синусов и косинусов

## <center> СИНУС, КОСИНУС, ТАНГЕНС, КОТАНГЕНС УГЛОВ ТРЕУГОЛЬНИКА

В прямоугольном треугольнике два острых угла. Их характеризуют такие величины, как синус, косинус, тангенс и котангенс. Ранее мы рассматривали эти функции с алгебраической точки зрения — теперь настало время посмотреть, как можно выразить значения этих функций через отношения сторон прямоугольного треугольника.

Пусть у нас есть треугольник $ABC$ с углом $\alpha$, для которого мы хотим выразить значения тригонометрических функций:

![](https://lms.skillfactory.ru/assets/courseware/v1/c4f71f7ef766c8255db639ac4217ae62/asset-v1:SkillFactory+DSPR-2.0+14JULY2021+type@asset+block/MFTIDSLIGHT_MATH_4_4_1.png)

- ***Синусом*** угла $\alpha$ называется отношение противолежащего катета к гипотенузе:

$\sin \alpha=\frac{AC}{AB}$

- ***Косинусом*** угла $\alpha$ называется отношение прилежащего катета к гипотенузе:

$\cos \alpha=\frac{BC}{AB}$

- ***Тангенсом*** угла $\alpha$ называется отношение противолежащего катета к прилежащему:

$\tg \alpha=\frac{AC}{BC}$

- ***Котангенсом*** угла $\alpha$ называется отношение прилежащего катета к противолежащему:

$\ctg \alpha=\frac{BC}{AC}$

Конечно, заучивать эти формулы необязательно, однако их знание сильно поможет в понимании математических выкладок. Чтобы их проще было запомнить, предлагаем вам обратиться к русскому языку: предлоги «к» и «ко» используется в значении приближения к чему-то. Так же и в формулах тригонометрических функций: если в названии функции есть в начале «ко» (косинус, котангенс), то мы начинаем вычисления с прилежащего катета (т. е. самого приближенного).

***Тригонометрические значения есть и у тупых углов***. Правила, по которым их можно найти, довольно простые:

1. Формула синуса тупого угла: $\sin (\alpha) = \sin(180˚ - \alpha)$

2. Формула косинуса тупого угла: $\cos (\alpha) = -\cos(180˚ - \alpha)$

3. Формулы тангенса и котангенса тупого угла:

$\tg (\alpha) = -\tg(180˚ - \alpha)$

$\ctg (\alpha) = -\ctg(180˚ - \alpha)$

Таким образом, чтобы найти тригонометрическое значение тупого угла, необходимо найти тригонометрическое значение острого угла, смежного с ним, и правильно поставить знак. К примеру, если мы хотим найти значение синуса для угла в 120°, то мы можем отметить, что 120=180-60, а синус для 60° нам известен:

$\sin(120)=\sin(180-60)=\sin(60)=\frac{\sqrt3}{2}$

Кроме того, что синусы и косинусы можно выразить из прямоугольного треугольника, их также можно выражать друг через друга, так как тригонометрические функции одного угла связаны рядом тригонометрических тождеств.

***ДЛЯ ЛЮБОГО УГЛА СПРАВЕДЛИВЫ СЛЕДУЮЩИЕ РАВЕНСТВА:***

1. $\cos^2\alpha+\sin^2\alpha=1$

2. $\tg^2\alpha+\ctg^2\alpha=1$

3. $1+\tg^2\alpha=\frac{1}{\cos^2\alpha}$

4. $1+\ctg^2\alpha=\frac{1}{\sin^2\alpha}$

Первая формула называется основным тригонометрическим тождеством. Пользуясь этими формулами, всегда можно найти все четыре тригонометрические величины, зная всего одну.

Рассмотрим пример. Пусть в прямоугольном треугольнике $ABC\cos(A)=0.6$. Необходимо найти все остальные тригонометрические функции этого угла.

В первую очередь мы можем найти синус по основному тригонометрическому тождеству:

$\cos^2\alpha+\sin^2\alpha=1$

Подставляем известный нам косинус и решаем уравнение, чтобы найти синус:

$(0.6)^2+sin^2(\alpha)=1$

$0.36+sin^2(\alpha)=1$

$sin^2(\alpha)=0.64$

$sin(\alpha)=0.8$

Теперь мы можем найти тангенс, поскольку он является отношением синуса к косинусу:

$\tg(A)=\frac{\sin(A)}{\cos(A)}=\frac{0.8}{0.6}=1\frac{1}{3}$

Котангенс же является отношением косинуса к синусу:

$\ctg(A)=\frac{\cos(A)}{\sin(A)}=\frac{0.6}{0.8}=\frac{3}{4}$


## <center> ТЕОРЕМА СИНУСОВ

Теорема синусов позволяет связать синусы углов и длины сторон в любом треугольнике. Согласно ей, стороны произвольного треугольника пропорциональны синусам противолежащих углов:

$\frac{a}{\sin\alpha}=\frac{b}{\sin\beta}=\frac{c}{\sin\gamma}$

Рассмотрим применение этой теоремы на примере. Пусть у нас есть треугольник, в котором известны два угла, а также известно, что $KN=4\sqrt6$:

![](https://lms.skillfactory.ru/assets/courseware/v1/0f82d43a34daefdf478c6968e1f4a260/asset-v1:SkillFactory+DSPR-2.0+14JULY2021+type@asset+block/MFTIDSLIGHT_MATH_4_4_3.png)

Нам необходимо найти длину стороны MN.

Запишем теорему синусов, используя те отношения, в которых нам известны углы:

$\frac{MN}{\sin K}=\frac{KN}{\sin M}$

Выразим искомую сторону:

$MN=\frac{KN\sin K}{\sin M}$

Подставим известные нам значения:

$MN=\frac{4\sqrt6\sin 30˚}{\sin 45˚}$

Получаем, что

$MN=4\sqrt3$

## <center> ТЕОРЕМА КОСИНУСОВ

Теорема косинусов гласит, что квадрат стороны треугольника равен сумме квадратов двух других сторон без удвоенного произведения этих сторон на косинус угла между ними:

$a^2=b^2+c^2-2*b*c*\cos\alpha$

К примеру, если нам известно, что в треугольнике есть две стороны, 6 и 10, а угол между ними равен 60°, то мы можем найти третью сторону:

$a^2=6^210^2-2*6*10*\cos60˚$


$a^2=36+100-120\cos60˚$

$a^2=136-60=76$

$a=\sqrt{76}$

# <center> MATH-4.2. Понятие угла. Параллельность и перпендикулярность прямых

## <center> УГЛЫ

***Угол*** — это геометрическая фигура, которая состоит из точки (вершины угла) и двух лучей (сторон угла), исходящих из этой точки.

![](https://lms.skillfactory.ru/assets/courseware/v1/03640e5545072f4134c8ace11f42ec86/asset-v1:SkillFactory+DSPR-2.0+14JULY2021+type@asset+block/MFTIDSLIGHT_MATH_4_2_1.png)

***Развёрнутый угол*** построен на двух полупрямых, дополняющих друг друг до прямой.

Это выглядит следующим образом:

![](https://lms.skillfactory.ru/assets/courseware/v1/9d6b3bdda8d87c6286c86cb73014444b/asset-v1:SkillFactory+DSPR-2.0+14JULY2021+type@asset+block/MFTIDSLIGHT_MATH_4_2_2.png)

Если развёрнутый угол разделить на 180 равных частей, каждая из этих частей будет называться ***градусом*** — единицей, в которой измеряют величину угла.

Классификация углов в зависимости от их градусной меры:

- Если угол равен 90°, его называют ***прямым***.
- Если угол меньше 90°, его называют ***острым***.
- Eсли угол больше 90°, но меньше 180°, его называют ***тупым***.


## <center> БИССЕКТРИСА

***Биссектрисой*** называют луч, выходящий из вершины угла и делящий этот угол на две равные части:

![](https://lms.skillfactory.ru/assets/courseware/v1/64af5d4f0887620357b519f45b7928cf/asset-v1:SkillFactory+DSPR-2.0+14JULY2021+type@asset+block/MFTIDSLIGHT_MATH_4_2_6.png)

На рисунке выше вы видите угол AOB и биссектрису OC. Так как луч OC является биссектрисой, то углы AOC и BOC равны.

## <CENTER> CМЕЖНЫЕ УГЛЫ

Два угла называются ***смежными***, если одна из их сторон является общей, а другие стороны дополняют друг друга до прямой.

![](https://lms.skillfactory.ru/assets/courseware/v1/1f69b9e8bc01f2a8367e8b85e1ab6423/asset-v1:SkillFactory+DSPR-2.0+14JULY2021+type@asset+block/MFTIDSLIGHT_MATH_4_2_8.png)

Проще говоря, два смежных угла образуют один развёрнутый, который как будто разделён «стеной» — отсюда и название. Так как смежные углы образуют один развёрнутый угол, то их сумма равна 180°.

Рассмотрим пример. Предположим, мы знаем, что один из двух смежных углов в четыре раза больше другого. Как найти величины обоих углов в градусах?

Подобные задачи удобно решать с помощью уравнений. Обозначим меньший угол за x (в плане вычислений всегда удобно обозначать через переменную наименьшую величину). Тогда больший угол будет равен 4x. Так как углы являются смежными и дают в сумме 180°, получаем следующее уравнение:

$x+4x=180$

$5x=180$

$x=36$

Меньший угол равен 36. Значит, больший угол равен 36*4=144.

## <center> ВЕРТИКАЛЬНЫЕ УГЛЫ

Два угла называются ***вертикальными***, если стороны одного угла дополняют лучи другого угла до прямой. Другими словами, вертикальные углы — это несмежные углы, которые образуются при пересечении двух прямых.

![](https://lms.skillfactory.ru/assets/courseware/v1/06340b8c9f1cbfa568ccea496563b9e5/asset-v1:SkillFactory+DSPR-2.0+14JULY2021+type@asset+block/MFTIDSLIGHT_MATH_4_2_9.png)

Главная особенность таких углов заключается в том, что они равны. То есть на рисунке выше углы AOB и COD являются вертикальными и равными. Углы AOC и BOD также вертикальные, а значит, равны.

## <center> ПЕРПЕНДИКУЛЯРНЫЕ ПРЯМЫЕ

Иногда при пересечении двух прямых получаются вертикальные углы, как на рисунке ниже:

![](https://lms.skillfactory.ru/assets/courseware/v1/8f5ece343f374ee488142aa24505a618/asset-v1:SkillFactory+DSPR-2.0+14JULY2021+type@asset+block/MFTIDSLIGHT_MATH_4_2_10.png)

Здесь все получившиеся углы — прямые (по 90°). В таком случае говорят, что прямые являются ***перпендикулярными***

Две пересекающиеся прямые называются ***перпендикулярными***, если они пересекаются под прямым углом.

Перпендикулярность прямых a и b обозначается как $a\bot b$.

***Секущая*** — это прямая, которая пересекает кривую в двух точках, а также прямая, пересекающая две другие прямые.

Пусть у нас есть прямые a и b, которые пересечены секущей c:

![](https://lms.skillfactory.ru/assets/courseware/v1/a54489d714a504e17d012dfc5a4d9ca6/asset-v1:SkillFactory+DSPR-2.0+14JULY2021+type@asset+block/MFTIDSLIGHT_MATH_4_2_11.png)

В таком случае:

- Пары $\angle4$ и $\angle6$, $\angle3$ и $\angle5$ называются ***односторонними*** углами.
- Пары $\angle4$ и $\angle5$, $\angle3$ и $\angle6$ называются ***накрест лежащими*** углами.
- Пары $\angle1$ и $\angle5$, $\angle2$ и $\angle6$, $\angle3$ и $\angle7$, $\angle4$ и $\angle8$ называются ***соответственными*** углами.

## <center> ПАРАЛЛЕЛЬНЫЕ ПРЯМЫЕ

***Параллельные прямые*** — это прямые, лежащие в одной плоскости и не имеющие точек пересечения.

***ПРИЗНАКИ ПАРАЛЛЕЛЬНОСТИ ПРЯМЫХ:***

1. Если при пересечении двух прямых секущей накрест лежащие углы равны, то прямые параллельны.
2. Если при пересечении двух прямых соответственные углы равны, то прямые параллельны.
3. Если при пересечении двух прямых секущей сумма односторонних углов равна 180°, то прямые параллельны.

***СВОЙСТВА ПАРАЛЛЕЛЬНЫХ ПРЯМЫХ:***

1. Если две параллельные прямые пересечены секущей, то накрест лежащие углы равны.
2. Если две параллельные прямые пересечены секущей, то соответственные углы равны.
3. Если две параллельные прямые пересечены секущей, то сумма односторонних углов равна 180°.

# <center> MATH-4.5. Многоугольники. Четырёхугольники. Окружности

***Многоугольник*** — это геометрическая фигура, ограниченная замкнутой ломаной линией, не имеющей точек самопересечения.



## <center> ЧЕТЫРЁХУГОЛЬНИКИ

Случайные четырёхугольники с разными сторонами и разными углами не очень нас интересуют, поскольку у них нет специальных свойств, которые были бы нам полезны. Однако есть ряд частных случаев четырёхугольников, которые мы рассмотрим далее:

- параллелограмм,
- ромб,
- прямоугольник,
- квадрат,
- трапеция.

У всех этих четырёхугольников есть одно общее свойство — сумма их углов равна 360°, поскольку они представляют собой два соединённых треугольника.

## <center> ПАРАЛЛЕЛОГРАММ

***Параллелограммом*** называется четырёхугольник, у которого противоположные стороны попарно параллельны:

![](https://lms.skillfactory.ru/assets/courseware/v1/3c2175126f361989c78a6ad409130ae0/asset-v1:SkillFactory+DSPR-2.0+14JULY2021+type@asset+block/MFTIDSLIGHT_MATH_4_5_4.png)

Параллелограмм обладает рядом свойств, которые отличают его от других четырёхугольников:

- У параллелограмма равны противоположные стороны и противоположные углы.
- Диагонали параллелограмма точкой пересечения делятся пополам.

![](https://lms.skillfactory.ru/assets/courseware/v1/a31e66a1d04429021135c2f503208cea/asset-v1:SkillFactory+DSPR-2.0+14JULY2021+type@asset+block/MFTIDSLIGHT_MATH_4_5_5.png)

Например, дан параллелограмм, стороны которого соотносятся как 2:3. Необходимо найти, чему равны стороны, если периметр (сумма всех сторон, P) параллелограмма равен 30.

Мы знаем, что стороны параллелограмма соотносятся как 2:3. Значит, меньшую сторону можем обозначить за 2x, а большую — за 3x. Так как противоположные стороны параллелограмма равны, то будет две стороны по 2x и две — по 3x. Так как периметр — это сумма всех сторон, можно составить уравнение:

$2x+2x+3x+3x=30$

$10x=30$

$x=3$

Значит, две меньшие стороны равны $2x=2*3=6$, а две большие — $3x*3=9$.

## <center> ПРЯМОУГОЛЬНИК

Так как прямоугольник является разновидностью параллелограмма, то для него верны все свойства последнего. Однако благодаря прямым углам у прямоугольника появляется ещё одно свойство: его диагонали всегда равны.

![](https://lms.skillfactory.ru/assets/courseware/v1/35180a87a7b6e5dcd37beba94d080f7c/asset-v1:SkillFactory+DSPR-2.0+14JULY2021+type@asset+block/MFTIDSLIGHT_MATH_4_5_6.png)

## <center> РОМБ

***Ромб*** тоже является частным случаем параллелограмма. Следовательно, для него справедливы все свойства последнего.

***Особенность ромба*** — равенство сторон, благодаря чему у него появляется дополнительное свойство: диагонали ромба взаимно перпендикулярны и в точке пересечения делятся пополам.

![](https://lms.skillfactory.ru/assets/courseware/v1/a029bd060730170f3d461013ee56705e/asset-v1:SkillFactory+DSPR-2.0+14JULY2021+type@asset+block/MFTIDSLIGHT_MATH_4_5_7.png)

Давайте решим задачу с использованием свойств ромба. Пусть у нас есть ромб ABCD и нам известно, что его сторона равна 5, а его бóльшая диагональ BD равна 8. Необходимо найти вторую диагональ.

Мы видим, что ромб делится диагоналями на четыре равных прямоугольных треугольника. Значит, мы можем воспользоваться теоремой Пифагора. BO — половина большей диагонали, то есть равняется 4. Значит, получаем:

$AO^2+BO^2=AB^2$

$AO^2+4^2=5^2$

$AO^2=9$

$AO=3$

Таким образом, половина искомой диагонали равна 3. Значит, вторая диагональ целиком равна $3*2=6$.

## <center> КВАДРАТ

***Квадрат*** — это параллелограмм, который одновременно является и ромбом, и прямоугольником.

Свойства квадрата совпадают со свойствами ромба и прямоугольника:

1. Все углы квадрата прямые.
2. Диагонали квадрата взаимно перпендикулярны, равны между собой и делят углы квадрата пополам.
3. Углы между диагоналями и сторонами равны 45°.

## <center> ТРАПЕЦИЯ

***Трапецией*** называется четырёхугольник, у которого две стороны параллельны, а две другие — нет. Параллельные стороны называются ***основаниями*** трапеции, а две другие стороны — ***боковыми сторонами***.

![](https://lms.skillfactory.ru/assets/courseware/v1/8aad1cb7b7519b8670cede888180755a/asset-v1:SkillFactory+DSPR-2.0+14JULY2021+type@asset+block/MFTIDSLIGHT_MATH_4_5_10.png)

У трапеции, как и у треугольника, есть средняя линия.

***Средней линией трапеции*** называется отрезок, соединяющий середины её боковых сторон.

Чтобы найти длину средней линии, необходимо сложить основания трапеции и поделить их сумму на 2.

![](https://lms.skillfactory.ru/assets/courseware/v1/ddbda7e20aa9b67cf1b6b5d52e738069/asset-v1:SkillFactory+DSPR-2.0+14JULY2021+type@asset+block/MFTIDSLIGHT_MATH_4_5_11.png)



## <center> ОКРУЖНОСТЬ

***Окружность*** — это множество всех точек плоскости, равноудалённых от точки, называемой ***центром***.

В окружности можно проводить отрезки, которые имеют собственные названия:

1. ***Радиус (R)*** — отрезок, соединяющий центр окружности и любую точку на ней.

2. ***Диаметр (D)*** — отрезок, соединяющий любые две точки на окружности и проходящий через центр.

![](https://lms.skillfactory.ru/assets/courseware/v1/a204468631b751c6283628bb4db9d580/asset-v1:SkillFactory+DSPR-2.0+14JULY2021+type@asset+block/MFTIDSLIGHT_MATH_4_5_13.png)

Диаметр в два раза больше радиуса: D=2*R.

3. ***Хорда*** — отрезок, соединяющий любые две точки на окружности.

![](https://lms.skillfactory.ru/assets/courseware/v1/11e9e5639fa83860cde0de5c9f22c16c/asset-v1:SkillFactory+DSPR-2.0+14JULY2021+type@asset+block/MFTIDSLIGHT_MATH_4_5_14.png)


Если начертить на плоскости окружность и прямую, их взаимное расположение может быть различным:
    
1. Прямая может не пересекать окружность ни в одной точке. Это происходит, если расстояние от центра окружности до прямой больше радиуса.

![](https://lms.skillfactory.ru/assets/courseware/v1/12f2a4810b92f83393078619be6c27d9/asset-v1:SkillFactory+DSPR-2.0+14JULY2021+type@asset+block/MFTIDSLIGHT_MATH_4_5_15.png)

2. Прямая может пересекать окружность в двух точках. Это происходит, когда расстояние от центра окружности до прямой меньше радиуса. В таком случае эту прямую называют ***секущей***.

![](https://lms.skillfactory.ru/assets/courseware/v1/9174d29692251ab548e6df7a6b3e722b/asset-v1:SkillFactory+DSPR-2.0+14JULY2021+type@asset+block/MFTIDSLIGHT_MATH_4_5_16.png)

3. Прямая может пересекать окружность в одной точке. Это происходит, когда расстояние от центра окружности до прямой равно радиусу. В таком случае эту прямую называют ***касательной***.

![](https://lms.skillfactory.ru/assets/courseware/v1/b8e7dba144edd0f86d05d4047626ddf6/asset-v1:SkillFactory+DSPR-2.0+14JULY2021+type@asset+block/MFTIDSLIGHT_MATH_4_5_17.png)

Часть окружности, заключённая между двумя точками, называется ***дугой***.

Заметим, что точки A и B делят изображённую ниже окружность на две дуги. Чтобы их различать, будем обозначать их тремя буквами в зависимости от того, какую точку содержит та или иная дуга. Итак, имеем две дуги: $\smile{ADB}$ и $\smile{ACB}$.

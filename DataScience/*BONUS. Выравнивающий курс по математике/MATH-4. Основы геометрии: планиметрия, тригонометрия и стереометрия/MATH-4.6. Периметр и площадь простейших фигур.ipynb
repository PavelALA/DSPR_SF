# <center> MATH-4.6. Периметр и площадь простейших фигур

## <center> ПЕРИМЕТР

***Периметр*** — это сумма длин всех сторон фигуры. Обычно его обозначают большой латинской буквой $P$. Также иногда встречается ***полупериметр*** — его обозначают маленькой латинской буквой $p$.

К примеру, если мы хотим найти периметр параллелограмма с большей стороной, равной 10, и меньшей стороной, равной 3, то нам нужно просто сложить длины всех сторон:

$P=10+3+10+3=26$

Чуть сложнее найти периметр у окружности. Если радиус окружности равен $R$, то её периметр (т. е. длина окружности) равен $2\pi R$.

# <center> ПЛОЩАДЬ

Площадью фигуры называют часть плоскости, ограниченную замкнутой линией. Площадь обычно обозначают заглавной латинской буквой $S$.

# <center> ПЛОЩАДЬ КВАДРАТА

Площадь квадрата со стороной $a$ вычисляется по формуле:

$S=a^2$



## <center> ПЛОЩАДЬ ПРЯМОУГОЛЬНИКА

Площадь прямоугольника со сторонами $a$ и $b$ вычисляется по формуле:

$S=a*b$



## <center> ПЛОЩАДЬ ТРЕУГОЛЬНИКА

Существует много формул для нахождения площади треугольника, однако мы рассмотрим только три из них, которые используются чаще всего.

1. $S=\frac{1}{2}ah$ - площадь треугольника равна полупроизведению стороны на высоту, опущенную на эту сторону

Пусть нам дан треугольник, основание которого равно 10, а высота, проведённая к нему, равна 5.

Тогда площадь этого треугольника можно найти следующим образом:

$S=\frac{1}{2}10*5=25$

2. $S=\frac{1}{2}ab*\sin\gamma$ - — площадь треугольника равна полупроизведению сторон на синус угла между ними.

Если у нас есть треугольник со сторонами 8 и 18 и углом между ними, равным 30°, то площадь такого треугольника будет равна:

$S=\frac{1}{2}*8*18*\frac{1}{2}=36$

3. $S=\sqrt{p(p-a)(p-b)(p-c)}$ - — формула Герона, где $p$ — полупериметр, а $a,b,c$  — стороны треугольника.

Если дан треугольник со сторонами 13, 14 и 15 и мы хотим найти его площадь, то сначала нужно найти полупериметр:

$p=\frac{13+14+15}{2}=21$

Затем мы можем просто вычислить площадь по формуле:

$S=\sqrt{21(21-13)(21-14)(21-15)}=84$

## <center> ПЛОЩАДЬ РОМБА

Ромб — это частный случай параллелограмма, поэтому его площадь можно находить через приведённые выше формулы. Однако есть дополнительная формула для поиска площади ромба — площадь равна полупроизведению диагоналей:

$S=\frac{d_1d_2}{2}$

## <center> ПЛОЩАДЬ ТРАПЕЦИИ

Чтобы найти площадь трапеции, необходимо умножить полусумму оснований на высоту трапеции:

$=\frac{a+b}{2}*h$

## <center> ПЛОЩАДЬ ОКРУЖНОСТИ (КРУГА)

Нам осталось научиться находить площадь окружности, вернее, площадь круга, так как под окружностью понимают именно линию, а не часть плоскости, ограниченную этой линией.

Площадь для круга с радиусом $R$ можно найти по следующей формуле:

$S=\pi R^2$

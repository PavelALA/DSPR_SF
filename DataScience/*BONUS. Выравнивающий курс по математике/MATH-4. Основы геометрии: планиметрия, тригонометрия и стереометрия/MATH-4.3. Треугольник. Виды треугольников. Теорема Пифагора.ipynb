# <center> MATH-4.3. Треугольник. Виды треугольников. Теорема Пифагора

***Треугольником*** называется геометрическая фигура, образованная тремя отрезками, которые соединяют три не лежащие на одной прямой точки. Эти точки называются ***вершинами*** треугольника, а отрезки — ***сторонами*** треугольника. Углы между лучами, содержащими две стороны и выходящими из одной вершины треугольника, называются ***углами*** треугольника.

![](https://lms.skillfactory.ru/assets/courseware/v1/7385520e2b149a2ee6fc7a8a058b5f50/asset-v1:SkillFactory+DSPR-2.0+14JULY2021+type@asset+block/MFTIDSLIGHT_MATH_4_3_1.png)

Первое важное для нас свойство треугольника — сумма углов любого треугольника всегда равна 180°. Таким образом, мы всегда можем найти величину третьего угла, если два угла нам известны.



## <center> ЧЕВИАНЫ

В треугольнике можно проводить разные отрезки, соединяющие вершину и противоположную сторону (или прямую, содержащую противоположную сторону). Такие отрезки называют обычно ***чевианами***. Некоторые из них имеют специальные названия:

1. ***Медиана*** — это отрезок, соединяющий вершину с серединой противолежащей стороны треугольника.

![](https://lms.skillfactory.ru/assets/courseware/v1/987eb91b592e6cbe80d7327de1dda89c/asset-v1:SkillFactory+DSPR-2.0+14JULY2021+type@asset+block/MFTIDSLIGHT_MATH_4_3_2.png)

2. ***Биссектриса*** — отрезок биссектрисы угла треугольника, соединяющий вершину с точкой на противолежащей стороне.

![](https://lms.skillfactory.ru/assets/courseware/v1/5f3abeb55f80c9b5098eb2d2fdb98939/asset-v1:SkillFactory+DSPR-2.0+14JULY2021+type@asset+block/MFTIDSLIGHT_MATH_4_3_3.png)

3. ***Высота*** — это перпендикуляр, проведённый из вершины к прямой, которая содержит противолежащую сторону треугольника. Важно обратить внимание, что высота может падать внутри треугольника, совпадать с одной из его сторон или падать вне треугольника:

![](https://lms.skillfactory.ru/assets/courseware/v1/ca98d8f6bafd187434f5cf773afe78dc/asset-v1:SkillFactory+DSPR-2.0+14JULY2021+type@asset+block/MFTIDSLIGHT_MATH_4_3_4.png)



## <center> СРЕДНЯЯ ЛИНИЯ ТРЕУГОЛЬНИКА

***Средней линией*** треугольника называется отрезок, соединяющий середины двух его сторон. К примеру, на иллюстрации ниже у треугольника ACD средней линией будет отрезок KB.

Средняя линия параллельна третьей стороне (в данном случае KB параллельна AD) и равна её половине.

## <center> ВИДЫ ТРЕУГОЛЬНИКОВ

Основная классификация треугольников происходит по их наибольшему углу:

- Если наибольший угол в треугольнике острый, то треугольник ***остроугольный***.
- Если наибольший угол в треугольнике прямой, то треугольник ***прямоугольный***.
- Если наибольший угол в треугольнике тупой, то треугольник ***тупоугольный***.

## <center> РАВНОБЕДРЕННЫЙ ТРЕУГОЛЬНИК

***Равнобедренный треугольник*** — это такой треугольник, у которого две стороны равны (такие стороны называются боковыми). В таком треугольнике также равны два угла при основании (третьей стороне), а медиана, высота и биссектриса, проведённые к основанию, совпадают:

## <center> РАВНОСТОРОННИЙ ТРЕУГОЛЬНИК

***Равносторонний треугольник*** — это треугольник, у которого все стороны и все углы равны. Разумеется, высота, медиана и биссектриса, проведённые к любой стороне, у него также совпадают.

![](https://lms.skillfactory.ru/assets/courseware/v1/59431123cdb5db587c21994a22c32f65/asset-v1:SkillFactory+DSPR-2.0+14JULY2021+type@asset+block/MFTIDSLIGHT_MATH_4_3_10.png)



## <center> ПРЯМОУГОЛЬНЫЙ ТРЕУГОЛЬНИК

***Прямоугольный треугольник*** — это треугольник, один из углов которого прямой (равен 90°). Сторона, которая лежит напротив угла в 90°, называется ***гипотенузой***, две другие стороны — ***катетами***.

![](https://lms.skillfactory.ru/assets/courseware/v1/4d8be4cc3b2dddc0855458eec5d55989/asset-v1:SkillFactory+DSPR-2.0+14JULY2021+type@asset+block/MFTIDSLIGHT_MATH_4_3_11.png)

У прямоугольного треугольника есть несколько важных ***свойств***, связанных с углами:

1. Сумма острых углов прямоугольного треугольника равна 90°.
2. Катет напротив угла в 30° в два раза меньше гипотенузы.
3. Если в прямоугольном треугольнике есть угол 45°, то такой треугольник равнобедренный и его катеты равны.

# <center> MATH-4.1. Введение. Базовые геометрические фигуры: точка, линия, отрезок, луч

В этом модуле мы познакомимся с основными понятиями геометрии. Геометрия существует в двух вариантах: на плоскости (***планиметрия***) и в пространстве (***стереометрия***). Мы начнём с планиметрии, а затем перейдём к стереометрии.

ЦЕЛИ МОДУЛЯ:

- узнать основные понятия из планиметрии и стереометрии;

- изучить базовые фигуры на плоскости и в пространстве;

- научиться решать задачи, используя полученные знания.

## <center> ТОЧКА

***Точка*** — это геометрическая фигура, из которой состоят все остальные геометрические фигуры. У неё нет формы и размера, она является неделимой. Это неопределяемый и неизмеримый объект, с которого начинается вся геометрия.

Единственная характеристика точки — её положение на плоскости или в пространстве. Каждая точка обычно обозначается заглавной буквой латинского алфавита:

## <center> ОТРЕЗОК

Если ограничить прямую с обеих сторон, получим ***отрезок***:

![](https://lms.skillfactory.ru/assets/courseware/v1/72b4ce2a4509bcc5794a1594b6838612/asset-v1:SkillFactory+DSPR-2.0+14JULY2021+type@asset+block/MFTIDSLIGHT_MATH_4_1_12.png)

Отрезок называют по точкам его окончания, их можно расположить в любом порядке. Таким образом, отрезок выше можно обозначить AB или BA.



## <center> ЛУЧ

Если на прямой отметить точку, то она разделит прямую на две части, у каждой из которых будет начало и не будет конца. Таким образом получится два луча.

***Луч*** — это часть прямой, ограниченная с одной стороны точкой.

Луч можно продолжить только в сторону, не ограниченную точкой.

![](https://lms.skillfactory.ru/assets/courseware/v1/1841f312cee1706e10fd98546fe74232/asset-v1:SkillFactory+DSPR-2.0+14JULY2021+type@asset+block/MFTIDSLIGHT_MATH_4_1_13.png)

К примеру, на рисунке выше луч ограничен точкой A, но не ограничен с другой стороны. При обозначении луча важен порядок точек: сначала необходимо обозначить начало луча, а затем — его конец. То есть луч выше можно назвать только AB. Направление луча обычно не обозначают с помощью стрелки.

## <center> ПРАВИЛО СУММЫ

***Правило суммы*** — это базовое правило, которое помогает вычислить количество вариантов без прямого перебора.

Предположим, мы хотим выбрать одного сотрудника, который будет представлять компанию на чемпионате. Мы можем выбрать его из одного из двух отделов: в первом работает 10 человек, а во втором — 6. Тогда общее количество вариантов сотрудников, которых мы можем отправить на чемпионат, равняется $10+6=16$.

Основная формулировка правила суммы гласит, что если есть $n$ вариантов одного действия и $m$ вариантов другого действия и эти два действия не могут быть выполнены одновременно, то существует $n+m$ способов выбрать одно из этих действий. Если записать это формально с точки зрения теории множеств, то:

Пусть есть два множества объектов $A={a_1,...,a_n}$ и $B={b_1,...,b_m}$, тогда количество способов выбрать один объект из A или один объект из B равно $n+m$.

Правило суммы применяется только к взаимоисключающим вариантам, то есть к случаям, когда можно выбрать элемент только из одного множества. Правило суммы следует применять, если задачу можно перефразировать, используя слово «ИЛИ». Например, в случае с чемпионатом это правило подошло, так как мы выбирали сотрудника, который работает ***или*** в первом отделе, ***или*** во втором — сотрудников, работающих в обоих отделах, нет.

Рассмотрим ещё один пример. Маргарита сегодня идёт на важную встречу, и ей нужно выбрать одну из трёх красных юбок ***или*** одну из четырёх синих юбок. Сколько вариантов выбрать юбку есть у Маргариты?

Поскольку Маргарита может надеть одну из трёх красных юбок ***или*** одну из четырёх синих юбок, эти варианты являются взаимоисключающими и к ним применимо правило суммы. Это даёт $3+4=7$ вариантов.

Как вы можете видеть, в каждой записи таблицы появляется разная комбинация вида кофе и размера стаканчика. Таким образом, общее количество возможных вариантов равно количеству записей в этой таблице. Для такой маленькой таблицы мы, конечно, можем просто подсчитать все записи и увидеть, что их двенадцать. Но также можно заметить, что количество записей в таблице равно количеству строк, умноженному на количество столбцов, то есть $4*3=12$. Другими словами, чтобы определить общее количество вариантов выбора, мы умножаем количество вариантов выбора типа кофе (количество строк) на количество вариантов выбора размера стаканчика (количество столбцов). Только что мы применили правило произведения. Сформулируем его в общем виде:

Пусть есть два множества объектов $A={a_1,...,a_n}$ и $B={b_1,...,b_m}$. Тогда количество способов выбрать сначала один объект из A, а затем — один объект из B равно $n*m$.

Рассмотрим ещё один пример. Допустим, номерной знак состоит из двух латинских букв и трёх цифр. Сколько существует вариантов номерных знаков?

Выбрать первую букву можно 26 способами, вторую — тоже. Также у нас есть 10 возможных вариантов для первой цифры, 10 — для второй и 10 — для третьей. Таким образом, всего существует $26^2+10^3=676000$ комбинаций.

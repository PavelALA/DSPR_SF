# <center> MATH-6.3. Перестановки

Допустим, вы ходите в фитнес-клуб, где за вами закреплён шкафчик, на котором надо самостоятельно задавать пин-код из цифр 2, 3, 4 и 5 таким образом, чтобы они не повторялись. Вы с удовольствием пользовались этой опцией, пока однажды не забыли пин-код. И подобрать его не так просто. Допустим, ваш код — это 2435. Если вы введёте 4325, шкафчик не откроется, так как порядок цифр отличается, т. е. 2435 и 4325 — это разные перестановки.

***Перестановка*** — это упорядоченный набор определённых элементов.

Вы решаете перебрать все комбинации, потому что это кажется вам более быстрым способом, чем обращаться за помощью к менеджеру. В целом, это правильное решение, так как вариантов действительно не очень много. Перестановками 2, 3, 4, 5 являются:

5432, 5423, 5324, 5342, 5234, 5243, 4532, 4523, 4325, 4352, 4253, 4235, 3542, 3524, 3425, 3452, 3254, 3245, 2543, 2534, 2435, 2453, 2354, 2345

Мы видим, что всего есть 24 варианта.

Что если мы хотим узнать, сколько существует вариантов перестановок, не перечисляя их, как это сделано выше? Например, мы хотим найти, сколько возможных перестановок четырёхзначных чисел можно составить из четырёх разных цифр.

Обозначим, что у нас есть четыре места, куда можно поставить цифры:

_ _ _ _

Первой цифрой может быть любая из четырёх цифр (то есть у нас есть четыре варианта, из которых мы выбираем), поэтому ставим 4 на первое место:

4 _ _ _

Теперь для второго места осталось три варианта, потому что мы уже использовали одну цифру. Так что поставим на вторую позицию 3:

4 3 _ _

Для третьей позиции осталось два варианта:

4 3 2 _

Для последней позиции осталось одно число, поэтому поставим там 1.

4 3 2 1

По принципу умножения мы знаем, что если существует x способов сделать первый выбор и y способов сделать второй выбор, то общее количество комбинаций равно $x*y$. Это означает, что для нахождения общего количества перестановок нам нужно просто перемножить количества вариантов для каждой позиции:

$4*3*2*1=24$

Намного удобнее использовать для записи такого выражения сокращённую факториальную нотацию (!):

***Факториалом*** числа называют произведение этого числа на все натуральные числа, которые меньше него. Факториал обозначается восклицательным знаком. Например, $6!=6*5*4*3*2*1=720!.

Количество перестановок обычно обозначают буквой $P$ с индексом, показывающим количество элементов:

$P_n=n!$

То есть количество перестановок из пяти элементов будет записано и вычислено следующим образом:

$P_5=5!=120$



Мы рассмотрели, как искать возможное количество перестановок, если повторения не предусмотрены. Но что, если нам нужно найти общее количество перестановок цифр 2, 3, 4 и 5, но мы также хотим также рассмотреть варианты вроде 2245 или 5555, где какие-то цифры не используются, а какие-то используются несколько раз? Как найти число таких перестановок?

Вычислить это даже проще. У нас вновь есть четыре позиции, на которые мы можем расставить возможное число вариантов:

Для первой позиции есть четыре варианта цифр, поэтому, как и раньше, ставим на первое место 4. Поскольку разрешено повторное использование цифр, на остальных позициях также будут стоять цифры 4:

4 4 4 4 = $4^4=256$

То есть количество перестановок из n элементов с повторами будет равно $n^n$.

Ещё один интересный для нас случай — это перестановки без повторений, когда среди элементов уже есть одинаковые. Предположим, что мы хотим найти количество последовательностей, переставляя буквы в слове «котик». Если бы все буквы были уникальными, количество перестановок было бы равно:

$P_5=5!=120$

Однако из-за одинаковых букв «к» перестановок получается в два раза меньше, потому что, к примеру, перестановка $к_1отик_2 и к_2отик_1$— это для нас одно и то же. Так что количество перестановок для слова «котик» будет равно:

$\frac{5!}{2}=60$

А что если одинаковых элементов три? Например, в слове «молоко» три буквы «о». Тогда абсолютно одинаковыми будут слова $мо_1ло_2ко_3$, $мо_1ло_3ко_2$, $мо_2ло_1ко_3$, $мо_2ло_3ко_1$, $мо_3ло_1ко_2$, $мо_3ло_2ко_1$. Каждое слово можно получить аж из шести различных перестановок. Значит, необходимо поделить количество перестановок на 6:

$\frac{6!}{6}=120

Однако 6 здесь получилось не просто так: это количество перестановок для трёх элементов (3!=6), ведь мы посчитали количество перестановок, которое можно получить из трёх букв «о».

То есть если есть повторы, необходимо поделить количество перестановок на количество перестановок для повторяющихся элементов. Формально это можно записать это следующим образом:

$P_{n_1,...n_k}=\frac{n!}{n_1!.....n_k!}$

$n=n_1+n_2+n_3+...+n_k$

Здесь через n с различными индексами обозначены группы одинаковых элементов (точнее, количество элементов в этих группах). Если элемент уникален, то он просто является группой с одним элементом.

Рассмотрим ещё один пример на применение этой формулы — слово «параллелограмм». В нём три раза повторяется буква «а», два раза — буква «р», три раза — буква «л» и два раза — буква «м». Нам обязательно стоит это учесть, ведь вне зависимости от того, поставим мы в начале первую «м» или вторую, слова для нас будут выглядеть одинаково. Поэтому мы должны поделить количество перестановок на количество перестановок каждой конкретной буквы:

$\frac{14!}{3!2!3!2!}=605404800$

# <center> MATH-6.7. Правила суммы и умножения вероятностей

## <center> ПРАВИЛО СУММЫ

Вероятностное ***правило суммы*** используется в ситуациях, когда необходимо найти вероятность наступления объединения событий.

К примеру, если есть события «Маша получила за тест 4» и «Маша получила за тест 5», их объединением будет событие «Маша получила за тест 4 или 5», и для вычисления его вероятности нам как раз понадобится правило суммы.

Это правило используется для несовместных событий, то есть событий, которые не могут произойти одновременно. Например, события «наступила ночь» и «наступил день» являются несовместными, так как в какой-то момент времени может происходить только одно из них.\

Если события A и B являются несовместными, то вероятность для объединения этих событий вычисляется по следующей формуле:

$P(A\cup B)=P(A)+P(B)$

Знак $\cup$ здесь имеет тот же смысл, что и в объединении множеств, то есть обозначает, что нас интересуют либо исходы для события A, либо исходы для события B.

Разумеется, формулу для вероятности объединения событий можно легко вывести математически. За S обозначим всё пространство исходов:

$P(A \cup B)=\frac{|A|+|B|}{|S|}=\frac{|A|}{|S|}+\frac{|B|}{|S|}=P(A)+P(B)$

Разберём это правило на примере:

В магазине одежды есть в наличии двое зелёных брюк, трое — красных и четверо — синих. Приходит покупатель и случайным образом выбирает брюки, причём любые брюки могут быть выбраны с одинаковой вероятностью. Точно известно, что он выберет только одни.

Какова вероятность того, что покупатель выберет зелёные или синие брюки?

Определим события G и B следующим образом:

- G = клиент выбирает зелёные брюки;
- B = клиент выбирает синие брюки.

$P(G\cup B)=P(G)+P(B)=\frac{2}{9}+\frac{4}{9}=\frac{2}{3}$

Получаем, что с вероятностью $\frac{2}{3}$ будут выбраны зелёные или синие брюки.

## <center> ОБОБЩЁННОЕ ПРАВИЛО СУММЫ

Применение рассмотренного нами правила суммы несколько ограничено, поскольку требует несовместных событий

Что, если события всё же могут быть совместными, то есть происходить одновременно?

Приведём обобщённое правило суммы, которое можно применять и в таких ситуациях:

$P(A\cup B)=P(A)+P(B)-P(A\cap B)$

![](https://lms.skillfactory.ru/assets/courseware/v1/a6b1e755d92f20c415dfebf2a9626e78/asset-v1:SkillFactory+DSPR-2.0+14JULY2021+type@asset+block/MATHML_md7_2_2.png)

Давайте рассмотрим, как выводится это правило.

Дело в том, что при сложении вероятностей наступления событий A и B мы считаем центральную часть (пересечение) дважды. Поэтому нам необходимо вычесть вероятность пересечения данных событий, чтобы она была вычислена только один раз.

$P(A\cup B)=\frac{|A|}{|S|}+\frac{|B|}{|S|}-\frac{|A\cap B|}{S}=|A|+|B|-|A\cap B|$

Давайте рассмотрим, как можно воспользоваться этим правилом, на примере:

Пусть 30% клиентов нашей клининговой компании — женщины. Также мы знаем, что 2/3 женщин пользуются нашими услугами еженедельной уборки, а 1/3 — нет. Среди мужчин 3/7 пользуются услугами еженедельной уборки, а 4/7 ими не пользуются.

Необходимо найти вероятность, что случайно выбранный клиент либо женского пола, либо пользуется нашими услугами еженедельной уборки, либо и то, и другое.

Просто сложить вероятность, что клиент женского пола, и вероятность, что клиент пользуется рассматриваемыми услугами, нельзя, так как в таком случае мы дважды посчитаем женщин, которые используют еженедельную уборку.

Можно представить наши данные в виде следующей таблицы:

- Польз услугами; Не польз услугами
- Женщины: 20%; 10%
- Мужчины: 30%; 40%

Чтобы рассчитать все значения, мы воспользовались информацией из условия задачи.

Если мы знаем, что всего женщин 30%, и 2/3 из них пользуются нашими услугами еженедельной уборки, а 1/3 — нет, то легко рассчитать:

- процент женщин, которые используют услугу: 30% * 2/3=20%;
- процент женщин, которые не используют услугу: 30% * 1/3=10$: 

Аналогичные расчёты можно провести и для мужчин.

Если пользоваться формулой, то для начала необходимо сложить вероятности того, что клиент женского пола: 20% + 10% = 30%

Затем надо прибавить вероятность того, что клиент пользуется нашими услугами: 30% + 20% = 50%

Тогда мы получим, что мы посчитали 30% и 10% по одному разу, а 20% — два раза:

Поэтому вычитаем 20% и получаем необходимый нам результат: 50% + 30% - 20% = 60%

***Задание 7.3***

В телекоммуникационной компании 34 тысячи клиентов. Из них 14 тысяч используют услугу домашнего интернета, 16 тысяч — услугу спутникового телевидения. Ещё 4 тысячи клиентов на данный момент не используют ни одну из перечисленных услуг.

Какова вероятность того, что случайно выбранный клиент использует домашний интернет или спутниковое телевидение?

$P(A \cup B)=\frac{14000}{34000}+\frac{16000}{34000}=0.882$

## <center> ПРАВИЛО УМНОЖЕНИЯ

Перейдём к правилу умножения, или, как его ещё называют, ***правилу произведения***.

***Правило произведения*** используется для нахождения вероятности пересечения событий:

$P(A\cap B)=P(A) * P(B)$

Рассмотрим его на примере:

Пусть нам известно, что среди людей, заходящих в магазин, 70 % ничего не покупают. Остальные с равной вероятностью покупают 1, 2 или 3 бутылки воды.

Допустим, что в магазин заходят два человека. Если событие A — это «покупка чётного числа бутылок первым человеком», а событие B — «покупка вторым человеком более чем одной бутылки», давайте найдём вероятность события $A\cap B$, то есть вероятность того, что один человек купил чётное количество бутылок, а второй купил более одной бутылочки.

Рассмотрим элементарные исходы, благоприятные событию A, и его вероятность:

$A{0,2}, P(A)=0.8$

Сделаем то же самое для события B:

$B{2,3}, P(B)=0.2$

Тогда вероятность их пересечения ищем следующим образом:

$P(A\cap B)=0.8 * 0.2= 0.16$

Какова вероятность того, что выбран король или карта червовой масти?

A - выбор короля B - выбор черви

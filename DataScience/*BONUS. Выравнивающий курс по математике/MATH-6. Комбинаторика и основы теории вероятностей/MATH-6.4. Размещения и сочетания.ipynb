# <center> MATH-6.4. Размещения и сочетания

## <center> РАЗМЕЩЕНИЯ

Представим, что есть 20 сотрудников, которые хотят участвовать в чемпионате. Нам необходимо выбрать капитана команды и его заместителя. То есть у нас есть всего два вакантных места: на первое будет 20 претендентов, а на второе — уже 19 (так как один человек уже займёт место капитана):

$\underline{20}*\underline{19}$

Мы, по сути, применили здесь правило умножения, и это абсолютно верно. Но что, если нам надо выбрать 10 человек? Тогда выражение будет уже таким:

$\underline{20}*\underline{19}*\underline{18}*\underline{17}*\underline{16}*\underline{15}*\underline{14}*\underline{13}*\underline{13}*\underline{12}*\underline{11}$

А если речь идёт о большем объёме данных? Выписывать 100 и больше множителей явно бессмысленно. Если мы хотим выбрать $k$ элементов из $n$, то в общем виде это можно записать следующим образом ($A^k_n$ — это количество размещений $k$ элементов из $n$ вариантов):

$A^k_n=n(n-1)(n-2)...(n-k+1)$

Умножаем числитель и знаменатель на $(n-k)!$:

$A^k_n=\frac {n(n-1)(n-2)...(n-k+1)(n-k)!}{n-k!}$

Упрощаем и получаем формулу размещений:

$A^k_n=\frac {n!}{n-k!}$

***Размещение из $n$ по $k$*** — это упорядоченный набор из $k$ различных элементов из некоторого множества различных элементов $n$.

Давайте применим формулу на примере. Допустим, в комнату вошли десять человек. Мы хотим посадить их на три стула, и порядок рассадки важен.

Тогда получаем следующее количество вариантов:

$A^k_n=\frac{10!}{7!}=10*9*8=720$

Мы рассмотрели случай так называемых размещений без повторений, иначе говоря, без возвращений — таких размещений, где мы не можем использовать выбранный элемент ещё раз. Однако можно составлять размещения и с повторениями. Например, если мы хотим составить комбинацию из трёх цифр, причём каждая цифра может использоваться любое количество раз, то число вариантов будет определяться по правилу умножения и равняться $3^3=27$.

Количество вариантов в случае размещений с повторениями вычисляется по следующей формуле:

$\overline{A^m_n}=m^n$

***Примечание***. Черта сверху обычно указывает на то, что элементы могут повторяться.

## <center> СОЧЕТАНИЯ

Вспомним кейс про сотрудников, из которых мы хотим собрать команду для чемпионата. Теперь попробуем понять, сколькими способами можно выбрать из них двух членов команды. Сперва может показаться, что задача абсолютно идентична выбору капитана команды и его заместителя: первого человека в команду мы можем выбрать двадцатью способами (потому что сотрудников 20), а второго — девятнадцатью. Значит, всего будет $20*19=380$ способов. Однако в данном случае это работает не совсем так, ведь пара «капитан и его заместитель» является упорядоченным набором элементов: назначить Марию капитаном, а Валентина — её заместителем — это совсем не то же самое, что назначить Валентина капитаном, а Марию — его заместителем. Если же говорить просто про двух членов команды, то совершенно не имеет значения, кого из них мы выбираем первым , а кого — вторым.

Представим, что в качестве двух участников мы выбираем некоторую неупорядоченную пару (Мария, Валентин). В эту неупорядоченную пару, по сути, «входят» две упорядоченные: (Мария, Валентин) и (Валентин, Мария). Иными словами, две разные пары дают один и тот же результат. Это значит, что число неупорядоченных пар будет в два раза меньше, чем число упорядоченных:

$\frac{20*19}{2}=190$

Теперь усложним задачу. Допустим, нам надо выбрать трёх сотрудников, то есть неупорядоченную тройку: (Мария, Валентин, Ольга). В эту неупорядоченную тройку «входят» шесть упорядоченных : (Мария, Валентин, Ольга), (Мария, Ольга, Валентин), (Валентин, Ольга, Мария), (Валентин, Мария, Ольга), (Ольга, Валентин, Мария), (Ольга, Мария, Валентин). В данном случае число неупорядоченных пар будет в шесть раз меньше, чем число упорядоченных:

$\frac{20*19*18}{6}=1140$

На самом деле мы могли не выписывать все возможные тройки, чтобы узнать, что их шесть, ведь это просто количество перестановок из трёх элементов, которое мы легко можем найти:

$P_n=P1=3!=3*2*1=6$

Таким образом, получается, что для того, чтобы найти количество сочетаний из $n$ элементов по $m$, надо поделить количество размещений из $n$ элементов по $m$ на количество перестановок из $m$ элементов:

$C^k_n=\frac{A^k_n}{P_k}$

Сочетанием из $n$ по $k$ называется набор из $k$ элементов, выбранных из множества объёмом $n$, в котором не учитывается порядок элементов.

Подставляем в числитель и знаменатель известные нам формулы и получаем:

$C^k_n=\frac{n(n-1)(n-2)...(n-k+1)}{k!}=\frac{n!}{k!(n-k)!}$

Рассмотрим пример применения формулы. Предположим, мы хотим найти количество способов, которыми можно выбрать пять фруктов из корзины, в которой всего 12 фруктов. Тогда, применяя формулу, получаем следующее:

$C_12^5=\frac{12!}{5!*7!}=792$

Теперь давайте усложним задачу: будем выбирать объекты не из одной группы, а из двух. Предположим, в компании работают две женщины и четверо мужчин. Мы хотим выбрать одну женщину и двоих мужчин для поездки в командировку. Сколькими способами мы можем это сделать?

Разумеется, порядок выбора для нас абсолютно не важен, однако нам важно, что мы выбираем один объект из двух (женщину) и два объекта из четырёх (мужчин) и далее комбинируем их.

Количество способов выбрать женщину будет следующим:

$C_2^1=\frac{2!}{1!*1!}=2$

Количество способов выбрать мужчин:

$C_4^2=\frac{4!}{2!*2!}=6$

Чтобы получить итоговый результат, воспользуемся правилом умножения:

$C_2^1*C_4^2=2*6=12$



Когда мы говорили о перестановках и размещениях, то рассматривали варианты с повтором и варианты без повтора. Разумеется, для сочетаний тоже можно найти количество способов выбрать элементы, если возможны повторения.

Число сочетаний с повторениями можно найти по следующей формуле:

$\overline{C^k_n}=C^k_{n+k-1}=\frac{(n+k-1)!}{k!(n-1)!}$

То есть, к примеру, если в задаче с выбором пяти фруктов из 12 мы можем возвращать фрукты и выбирать их заново, то количество вариантов изменится:

$\overline{C^5_12}=C^5_(12+5-1)=\frac{(12+5-1)!}{5!(12-1)!}=\frac{16!}{5!*11!}=4368$

***Задание 4.7***

Мария пришла в магазин и хочет выбрать 3 яблока и 2 апельсина. Всего на прилавке лежит 10 яблок и 6 апельсинов. Сколькими способами она может это сделать?

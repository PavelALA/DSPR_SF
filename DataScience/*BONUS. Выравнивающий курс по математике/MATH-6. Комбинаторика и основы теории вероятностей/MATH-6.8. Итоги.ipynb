Задание 8.1

Вы хотите сформировать бизнес-ланч из предложенного в кафе ассортимента. Можно выбрать один из трёх супов, одно из двух вторых блюд, а также один из четырёх напитков.

Сколько разных вариантов обеда вы можете заказать?

Воспользуемся правилом умножения:

- Три варианта первого блюда.
- Два варианта второго блюда.
- Четыре варианта напитка.

Получаем: 4*3*2=24.

Задание 8.4

Сколькими способами можно выбрать двух человек из 12 соискателей, пришедших на собеседование?

$C_12^2=\frac{12!}{2!*10!}=66$

Задание 8.8

Вероятность того, что заявление на оформление налогового вычета примут с первого раза, равна 0.8. Вы идёте подавать заявления вместе с другом.

Какова вероятность того, что хотя бы у одного из вас примут заявление?

1. Первый способ решения — от обратного: вычтем из единицы вероятность того, что заявление не примут ни у кого (обратное событие к интересующему нас событию):

1-0.2*0.2=0.96

2. Второй способ:

$P(A\cup B)=P(A)+P(B)-P(A\cap B)=0.8+0.8-(0.8*0.8)=0.96$
